### Question

(ex. 11 (b), p. 31, sec. 1.3, Kolman) Compute $BA$ where
\begin{align}
A = \begin{bmatrix}
1 & 2 & 3 \\
2 & 1 & 4
\end{bmatrix}
\quad
B =
\begin{bmatrix}
1 & 0  \\
2 & 1 \\
3 & 2
\end{bmatrix}
\end{align}

#### Solution

\begin{align}
BA &= \begin{bmatrix}
1 \cdot 1 + 0 \cdot 2 & 1 \cdot 2 + 0 \cdot 1 & 1 \cdot 3 + 0 \cdot 4 \\
2 \cdot 1 + 1 \cdot 2 & 2 \cdot 2 + 1 \cdot 1 & 2 \cdot 3 + 1 \cdot 4 \\
3 \cdot 1 + 2 \cdot 2 & 3 \cdot 2 + 2 \cdot 1 & 3 \cdot 3 + 2 \cdot 4
\end{bmatrix} \\
&= \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 10 \\
7 & 8 & 17
\end{bmatrix}
\end{align}

### Question

(ex. 1 (b), p. 113, sec. 2.2, Kolman) Solve the following linear system using _Gaussian Elimination_.

Note. _reduced row echelon form_ satisfies conditions (a), (b), (c), and (d). _Echelon form_ satisfies (a), (b), and (c).

\begin{align}
\left [
\begin{array}{cccc|c}
        1 & -3 & 4 & 1 & 0 \\
        0 & 0 & 1 & -1 & 4 \\
        0 & 0 & 0 & 1 & 1
\end{array}
\right ]
\end{align}

#### Solution

The system is already in echelon form.

\begin{align}
x - 3y + 4z + w &= 0 \\
z - w &= 4 \\
w &= 1
\end{align}

\begin{align}
w &= 1 \\
z &= 5 \\
x &= 3y - 21
\end{align}

Solution is
\begin{align}
\begin{bmatrix} x \\ y \\ z \\ w \end{bmatrix}
=
\begin{bmatrix}
3y - 21 \\ y \\ 5 \\ 1
\end{bmatrix}
\end{align}
Where $y$ is a free-variable and $x$ is dependent on it.

### Question

(ex. 3, p. 164, sec. 3.3, Kolman) Find the determinant by cofactor method of

\begin{bmatrix}
-1 & 2 & 3 \\
-2 & 5 & 4 \\
0 & 1 & -3
\end{bmatrix}

#### Solution

On the third row:
\begin{align}
&(-1)^4 (0)
\begin{vmatrix}
2 & 3 \\
5 & 4
\end{vmatrix}
+
(-1)^5 (1)
\begin{vmatrix}
-1 & 3 \\
-2 & 4
\end{vmatrix}
+
(-1)^6 (-3)
\begin{vmatrix}
-1 & 2 \\
-2 & 5
\end{vmatrix} \\
&= 0 - (-4 - (-6)) + (-3)(-5 - (-4)) \\
&= -2 + 3 = 1
\end{align}

### Question

(ex. 2, p. 164, sec. 3.5, Kolman) Solve the following system using _Cramer_.

\begin{align}
\begin{bmatrix}
1 & 1 & 1 & -2 \\
0 & 2 & 1 & 3 \\
2 & 1 & -1 & 2 \\
1 & -1 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
x_1 \\ x_2 \\ x_3 \\ x_4
\end{bmatrix}
=
\begin{bmatrix}
-4 \\ 4 \\ 5 \\ 4
\end{bmatrix}
\end{align}
Given
\begin{align}
\det A &= \det
\begin{bmatrix}
1 & 1 & 1 & -2 \\
0 & 2 & 1 & 3 \\
2 & 1 & -1 & 2 \\
1 & -1 & 0 & 1
\end{bmatrix}
= 26 \\ \\
\det A_1 &= \det
\begin{bmatrix}
-4 & 1 & 1 & -2 \\
4 & 2 & 1 & 3 \\
5 & 1 & -1 & 2 \\
4 & -1 & 0 & 1
\end{bmatrix}
= -52 \\
\det A_2 &= \det
\begin{bmatrix}
1 & -4 & 1 & -2 \\
0 & 4 & 1 & 3 \\
2 & 5 & -1 & 2 \\
1 & 4 & 0 & 1
\end{bmatrix}
= 0 \\ \\
\det A_3 &= \det
\begin{bmatrix}
1 & 1 & -4 & -2 \\
0 & 2 & 4 & 3 \\
2 & 1 & 5 & 2 \\
1 & -1 & 4 & 1
\end{bmatrix}
= 26 \\ \\
\det A_4 &= \det
\begin{bmatrix}
1 & 1 & 1 & -4 \\
0 & 2 & 1 & 4 \\
2 & 1 & -1 & 5 \\
1 & -1 & 0 & 4
\end{bmatrix}
= 26
\end{align}

#### Solution

\begin{align}
x_1 &= \frac{det \, A_1}{det \, A} = \frac{-30}{-30} = 1 \\
x_2 &= \frac{det \, A_2}{det \, A} = \frac{30}{-30} = -1 \\
x_3 &= \frac{det \, A_3}{det \, A} = \frac{0}{-30} = 0 \\
x_4 &= \frac{det \, A_4}{det \, A} = \frac{-60}{-30} = 2
\end{align}

### Question

(ex. 4, p. 169, sec. 3.4, Kolman) Find the inverse using the adjoint matrix method.

\begin{bmatrix}
2 & 1 & 3 \\
-1 & 2 & 0 \\
3 & -2 & 1
\end{bmatrix}

#### Solution

Minors.

\begin{align}
det(M_{11}) &= \begin{vmatrix} 2 & 0 \\ -2 & 1 \end{vmatrix} = 2 - 0 = 2 \\
det(M_{12}) &= \begin{vmatrix} -1 & 0 \\ 3 & 1 \end{vmatrix} = -1 - 0 = -1 \\
det(M_{13}) &= \begin{vmatrix} -1 & 2 \\ 3 & -2 \end{vmatrix} = 2 - 6 = -4 \\
det(M_{21}) &= \begin{vmatrix} 1 & 3 \\ -2 & 1 \end{vmatrix} = 1 - (-6) = 7 \\
det(M_{22}) &= \begin{vmatrix} 2 & 3 \\ 3 & 1 \end{vmatrix} = 2 - 9 = -7 \\
det(M_{23}) &= \begin{vmatrix} 2 & 1 \\ 3 & -2 \end{vmatrix} = -4 - 3 = -7 \\
det(M_{31}) &= \begin{vmatrix} 1 & 3 \\ 2 & 0 \end{vmatrix} = 0 - 6 = -6 \\
det(M_{32}) &= \begin{vmatrix} 2 & 3 \\ -1 & 0 \end{vmatrix} = 0 - (-3) = 3 \\
det(M_{33}) &= \begin{vmatrix} 2 & 1 \\ -1 & 2 \end{vmatrix} = 4 - (-1) = 5 \\
\end{align}

Cofactors

\begin{align}
A_{11} = (-1)^{1+1} det(M_{11}) = 1 \cdot 2 = 2 \\
A_{12} = (-1)^{1+2} det(M_{12}) = -1 \cdot -1 = 1 \\
A_{13} = (-1)^{1+3} det(M_{13}) = 1 \cdot -4 = -4 \\
A_{21} = (-1)^{2+1} det(M_{21}) = -1 \cdot 7 = -7 \\
A_{22} = (-1)^{2+2} det(M_{22}) = 1 \cdot -7 = -7 \\
A_{23} = (-1)^{2+3} det(M_{23}) = -1 \cdot -7 = 7 \\
A_{31} = (-1)^{3+1} det(M_{31}) = 1 \cdot -6 = -6 \\
A_{32} = (-1)^{3+2} det(M_{32}) = -1 \cdot 3 = -3 \\
A_{33} = (-1)^{3+3} det(M_{33}) = 1 \cdot 5 = 5
\end{align}

Determinant

$$a_{13} A_{13} + a_{23} A_{23} + a_{33} A_{33} = (3 \cdot -4 )+ 0 + (1 \cdot 5) = -7$$

Adjoint Matrix

$$
adj \, A =
\begin{bmatrix}
2 & -7 & -6 \\
1 & -7 & -3 \\
-4 & 7 & 5
\end{bmatrix}
$$

Inverse Matrix

$$
A^{-1} = \frac{1}{det \, A} \cdot adj \, A =
\begin{bmatrix}
\frac{-2}{7} & 1 & \frac{6}{7} \\
\frac{-1}{7} & 1 &  \frac{3}{7} \\
\frac{-4}{7} & -1 & \frac{-5}{7}
\end{bmatrix}
$$
