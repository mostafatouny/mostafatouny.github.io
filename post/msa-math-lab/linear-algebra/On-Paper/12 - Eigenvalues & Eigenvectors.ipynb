### Question

(ex. 6 (b), sec. 7.1, p. 450, Briggs) Find the characteristic polynomial, Eigenvalues, and Eigenvectors of

$$
\begin{bmatrix}
1 & 0 & 0 \\
-1 & 3 & 0 \\
3 & 2 & -2
\end{bmatrix}
$$

#### Solution

The characteristic polynomial

\begin{align}
\begin{vmatrix}
1 - \lambda & 0 & 0 \\
-1 & 3 - \lambda & 0 \\
3 & 2 & -2 - \lambda
\end{vmatrix}
&=
(1-\lambda) \big [ 0 - (3 - \lambda) (-\lambda - 2) \big ]
=
p(\lambda)
\end{align}

Eigenvalues

\begin{align}
0 = p(\lambda) &= (1-\lambda) (-1) (3 - \lambda) (-\lambda - 2) \\
&=  (1-\lambda) (3 - \lambda) (-\lambda - 2) \\
\lambda_1 &= 1 \\
\lambda_2 &= 3 \\
\lambda_3 &= -2
\end{align}

Eigenvectors of $\lambda_3$ are solution vectors of

$$
\begin{bmatrix}
1 & 0 & 0 \\
-1 & 3 & 0 \\
3 & 2 & -2
\end{bmatrix}
\begin{bmatrix}
x_1 \\ x_2 \\ x_3
\end{bmatrix}
=
-2 \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
$$

Yielding
\begin{align}
3x_1 + 0x_2 = 0 \\
-x_1 + 5x_2 = 0 \\
3x_1 + 2x_2 = 0
\end{align}

For example $v_3 = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}$ is an eigenvector corresponding to $\lambda_3$.
