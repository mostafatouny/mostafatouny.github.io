### Question

(ex. 11 (a), p. 31, sec. 1.3, Kolman) Compute $AB$ where
\begin{align}
A = \begin{bmatrix}
1 & 2 & 3 \\
2 & 1 & 4
\end{bmatrix}
\quad
B =
\begin{bmatrix}
1 & 0  \\
2 & 1 \\
3 & 2
\end{bmatrix}
\end{align}

#### Solution

\begin{align}
AB &= \begin{bmatrix}
1 \cdot 1 + 2 \cdot 2 + 3 \cdot 3 & 1 \cdot 0 + 2 \cdot 1 + 3 \cdot 2 \\
2 \cdot 1 + 1 \cdot 2 + 4 \cdot 3 & 2 \cdot 0 + 1 \cdot 1 + 4 \cdot 2
\end{bmatrix} \\
&= \begin{bmatrix}
1 + 4 + 9 & 0 + 2 + 6 \\
2 + 2 + 12 & 0 + 1 + 8
\end{bmatrix} \\
&= \begin{bmatrix}
14 & 8 \\
16 & 9
\end{bmatrix}
\end{align}

**P.S.** See blue-box in p. 144 (Kolman) for remarks about determinants.

### Question

(ex. 1 (a), p. 113, sec. 2.2, Kolman) Solve the following linear system using _Gaussian Elimination_.

Note. _reduced row echelon form_ satisfies conditions (a), (b), (c), and (d). _Echelon form_ satisfies (a), (b), and (c).

\begin{align}
\left [
\begin{array}{ccc|c}
        1 & 2 & -1 & 6 \\
        0 & 1 & 1 & 5 \\
        0 & 0 & 1 & 4
\end{array}
\right ]
\end{align}

#### Solution

The matrix is already in _echelon form_.
\begin{align}
x + 2y - z &= 6 \\
y + z &= 5 \\
z &= 4
\end{align}

\begin{align}
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
8 \\
1 \\
4
\end{bmatrix}
\end{align}

### Question

\begin{align}
det \begin{bmatrix}
2 & 0 & 0 \\
0 & -3 & 0 \\
0 & 0 & 4
\end{bmatrix}
\end{align}

#### Hint

#### Solution

### Question

(ex. 1, p. 164, sec. 3.3, Kolman) Find the determinant by cofactor method of

\begin{bmatrix}
1 & 0 & -2 \\
3 & 1 & 4 \\
5 & 2 & -3
\end{bmatrix}

On the first row:
\begin{align}
&(-1)^2 (1) \begin{vmatrix}
1 & 4 \\
2 & -3
\end{vmatrix}
+ 0 +
(-1)^4 (-2) \begin{vmatrix}
3 & 1 \\
5 & 2
\end{vmatrix} \\
&= (-3) - 8 + (-2)(6 - 5) \\
&= -11 - 2 = -13
\end{align}

### Question

(ex. 1, p. 164, sec. 3.5) Solve the following system using _Cramer_.

\begin{align}
2x_1 + 4x_2 + 6x_3 &= 2 \\
x_1 + 0x_2 + 2x_3 &= 0 \\
2x_1 +3x_2 -x_3 &= -5
\end{align}
Given
\begin{align}
\det A &= \det
\begin{bmatrix}
        2 & 4 & 6 \\
        1 & 0 & 2 \\
        2 & 3 & -1
\end{bmatrix}
= 26 \\
\det A_1 &= \det
\begin{bmatrix}
        2 & 4 & 6 \\
        0 & 0 & 2 \\
        -5 & 3 & -1
\end{bmatrix}
= -52 \\
\det A_2 &= \det
\begin{bmatrix}
        2 & 2 & 6 \\
        1 & 0 & 2 \\
        2 & -5 & -1
\end{bmatrix}
= 0 \\
\det A_3 &= \det
\begin{bmatrix}
        2 & 4 & 2 \\
        1 & 0 & 0 \\
        2 & 3 & -5
\end{bmatrix}
= 26 \\
\end{align}

#### Solution

\begin{align}
x_1 &= \frac{det \, A_1}{det \, A} = \frac{-52}{26} = -2 \\
x_2 &= \frac{det \, A_2}{det \, A} = \frac{0}{26} = 0 \\
x_3 &= \frac{det \, A_3}{det \, A} = \frac{26}{26} = 1
\end{align}
