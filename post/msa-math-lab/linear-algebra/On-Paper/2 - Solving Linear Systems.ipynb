### Question

(Source is linked below in the case-study). Solve the following system of linear equations, and state its number of solutions.

\begin{align}
1x + 1y &= 32 \\
5x + 10y &= 220
\end{align}

#### Solution

\begin{align}
\left [
\begin{array}{cc|c}
        1 & 1 & 32 \\
        5 & 10 & 220 \\
    \end{array}
\right ]
\rightarrow^{R_2 = R_2/5}
\left [
\begin{array}{cc|c}
        1 & 1 & 32 \\
        1 & 2 & 44 \\
    \end{array}
\right ]
\rightarrow^{R_2 = R_2 + (-1)R_1}
\left [
\begin{array}{cc|c}
        1 & 1 & 32 \\
        0 & 1 & 12 \\
    \end{array}
\right ]
\rightarrow^{R_1 = R_1 + (-1)R_2}
\left [
\begin{array}{cc|c}
        1 & 0 & 20 \\
        0 & 1 & 12 \\
    \end{array}
\right ]
\end{align}

In another notation:
\begin{align}
\begin{bmatrix}
        1 & 0 \\
        0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
        x \\
        y
\end{bmatrix}
&=
\begin{bmatrix}
        20 \\
        12
\end{bmatrix}
\end{align}

Concluding $x = 20$ and $y = 12$. A unique solution.

\begin{align}
1x + 0y = 20 \\
0x + 1y = 12
\end{align}

### Question

(ex. 8 (a), sec. 2.2, kolman). Solve the linear system



$$
\left [
\begin{array}{cccc|c}
        1 & 2 & 3 & 1 & 8 \\
        1 & 3 & 0 & 1 & 7 \\
        1 & 0 & 2 & 1 & 3
\end{array}
\right ]
$$

#### Solution

$$
\left [
\begin{array}{cccc|c}
        1 & 2 & 3 & 1 & 8 \\
        1 & 3 & 0 & 1 & 7 \\
        1 & 0 & 2 & 1 & 3
\end{array}
\right ]
\rightarrow^{R_2 = -R_1 + R_2}
\left [
\begin{array}{cccc|c}
        1 & 2 & 3 & 1 & 8 \\
        0 & 1 & -3 & 0 & -1 \\
        1 & 0 & 2 & 1 & 3
\end{array}
\right ]
\rightarrow^{R_3 = -R_1 + R_3}
\left [
\begin{array}{cccc|c}
        1 & 2 & 3 & 1 & 8 \\
        0 & 1 & -3 & 0 & -1 \\
        0 & -2 & -1 & 0 & -5
\end{array}
\right ]
\rightarrow^{R_1 = -2 R_2 + R_1}
\left [
\begin{array}{cccc|c}
        1 & 0 & 9 & 1 & 10 \\
        0 & 1 & -3 & 0 & -1 \\
        0 & -2 & -1 & 0 & -5
\end{array}
\right ]
\rightarrow^{R_3 = 2 R_2 + R_3}
\left [
\begin{array}{cccc|c}
        1 & 0 & 9 & 1 & 10 \\
        0 & 1 & -3 & 0 & -1 \\
        0 & 0 & -7 & 0 & -7
\end{array}
\right ]
\rightarrow^{R_3 = R_3/-7}
\left [
\begin{array}{cccc|c}
        1 & 0 & 9 & 1 & 10 \\
        0 & 1 & -3 & 0 & -1 \\
        0 & 0 & 1 & 0 & 1
\end{array}
\right ]
\rightarrow^{R_2 = 3 R_3 + R_2}
\left [
\begin{array}{cccc|c}
        1 & 0 & 9 & 1 & 10 \\
        0 & 1 & 0 & 0 & 2 \\
        0 & 0 & 1 & 0 & 1
\end{array}
\right ]
\rightarrow^{R_1 = -9 R_3 + R_1}
\left [
\begin{array}{cccc|c}
        1 & 0 & 0 & 1 & 1 \\
        0 & 1 & 0 & 0 & 2 \\
        0 & 0 & 1 & 0 & 1
\end{array}
\right ]
$$
Therefore $z = 1$, $y = 2$, and $x + m = 1$. So the solution is
\begin{align}
\begin{bmatrix} x \\ y \\ z \\ m \end{bmatrix}
=
\begin{bmatrix} 1-m \\ 2 \\ 1 \\ m \end{bmatrix}
\end{align}
Where $m$ is a free variable.

### Question

(ex. 8 (b), sec. 2.2, kolman). Solve the linear system
$$
\left [
\begin{array}{cccc|c}
        1 & 1 & 3 & -3 & 0 \\
        0 & 2 & 1 & -3 & 3 \\
        1 & 0 & 2 & -1 & -1
\end{array}
\right ]
$$

#### Solution

$$
\left [
\begin{array}{cccc|c}
        1 & 1 & 3 & -3 & 0 \\
        0 & 2 & 1 & -3 & 3 \\
        1 & 0 & 2 & -1 & -1
\end{array}
\right ]
\rightarrow^{R_3 = -R_1 + R_3}
\left [
\begin{array}{cccc|c}
        1 & 1 & 3 & -3 & 0 \\
        0 & 2 & 1 & -3 & 3 \\
        0 & -1 & -1 & 2 & -1
\end{array}
\right ]
\rightarrow^{R_2 = \frac{1}{2} R_2}
\left [
\begin{array}{cccc|c}
        1 & 1 & 3 & -3 & 0 \\
        0 & 1 & \frac{1}{2} & -\frac{3}{2} & \frac{3}{2} \\
        0 & -1 & -1 & 2 & -1
\end{array}
\right ]
\rightarrow^{R_1 = -R_2 + R_1}
\left [
\begin{array}{cccc|c}
        1 & 0 & \frac{5}{2} & -\frac{3}{2} & -\frac{3}{2} \\
        0 & 1 & \frac{1}{2} & -\frac{3}{2} & \frac{3}{2} \\
        0 & -1 & -1 & 2 & -1
\end{array}
\right ]
\rightarrow^{R_3 = R_2 + R_3}
\left [
\begin{array}{cccc|c}
        1 & 0 & \frac{5}{2} & -\frac{3}{2} & -\frac{3}{2} \\
        0 & 1 & \frac{1}{2} & -\frac{3}{2} & \frac{3}{2} \\
        0 & 0 & -\frac{1}{2} & \frac{1}{2} & \frac{1}{2}
\end{array}
\right ]
\rightarrow^{R_3 = -2 R_3}
\left [
\begin{array}{cccc|c}
        1 & 0 & \frac{5}{2} & -\frac{3}{2} & -\frac{3}{2} \\
        0 & 1 & \frac{1}{2} & -\frac{3}{2} & \frac{3}{2} \\
        0 & 0 & 1 & -1 & -1
\end{array}
\right ]
\rightarrow^{R_1 = -\frac{5}{2} R_3 + R_1}
\left [
\begin{array}{cccc|c}
        1 & 0 & 0 & 1 & 1 \\
        0 & 1 & \frac{1}{2} & -\frac{3}{2} & \frac{3}{2} \\
        0 & 0 & 1 & -1 & -1
\end{array}
\right ]
\rightarrow^{R_2 = -\frac{1}{2} R_3 + R_2}
\left [
\begin{array}{cccc|c}
        1 & 0 & 0 & 1 & 1 \\
        0 & 1 & 0 & -1 & 2 \\
        0 & 0 & 1 & -1 & -1
\end{array}
\right ]
$$
Therefore
\begin{align}
z - m &= -1 &\rightarrow &z = m - 1\\
y - m &= 2 &\rightarrow &y = m + 2\\
x + m &= 1 &\rightarrow &x = -m + 1
\end{align}
So the solution is
\begin{align}
\begin{bmatrix}
x \\
y \\
z \\
m
\end{bmatrix}
=
\begin{bmatrix}
-m + 1 \\
m + 2 \\
m - 1 \\
m
\end{bmatrix}
\end{align}
Where $m$ is a free variable.
