### Factorization

**Question.** Factorize the following algebraic form:
\begin{align}
a^2 - b^2
\end{align}

**Answer.**
\begin{align}
a^2 - b^2 = a^2 - ab + ab - b^2 = a(a-b) + b(a-b) = (a+b)(a-b)
\end{align}

### Recurrence

**Question.** Let $f(n) = n$ on the domain of natural numbers. Design a recursive function $g$ where $g(n) = f(n)$.

**Answer.**
\begin{align}
g(0) &= 0 \\
g(n) &= g(n-1) + 1 \qquad n \geq 1
\end{align}
