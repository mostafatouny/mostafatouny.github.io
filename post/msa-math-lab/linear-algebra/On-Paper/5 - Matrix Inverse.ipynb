### Question

(ex. 7, p. 124, sec. 2.3, Kolman) Find the inverse of by the procedure of page 121

$$
A =
\begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix}
$$

#### Solution

\begin{align}
\left [
A \, \bigg | \, I_2
\right ]
&=
\left [
\begin{array}{cc|cc}
        1 & 3 & 1 & 0 \\
        2 & 4 & 0 & 1 \\
    \end{array}
\right ] \\
&\rightarrow^{R_2 = -2 R_1 + R_2}
\left [
\begin{array}{cc|cc}
        1 & 3 & 1 & 0 \\
        0 & -2 & -2 & 1 \\
    \end{array}
\right ] \\
&\rightarrow^{R_2 = (1/-2) R_2}
\left [
\begin{array}{cc|cc}
        1 & 3 & 1 & 0 \\
        0 & 1 & 1 & \frac{-1}{2} \\
    \end{array}
\right ] \\
&\rightarrow^{R_1 = -3 R_2 + R_1}
\left [
\begin{array}{cc|cc}
        1 & 0 & -2 & \frac{3}{2} \\
        0 & 1 & 1 & \frac{-1}{2} \\
    \end{array}
\right ] \\
&=
\left [
I_2 \, \bigg | \, A^{-1}
\right ]
\end{align}

### Question

Given

\begin{align}
A = \begin{bmatrix} 1 & 3 \\ 2 & 4 \end{bmatrix} \quad
A^{-1} = \begin{bmatrix} -2 & \frac{3}{2} \\ 1 & \frac{-1}{2} \end{bmatrix} \quad
B = \begin{bmatrix} 7 \\ 7 \end{bmatrix}
\end{align}

What is the solution of the system
$$
AX = B
$$

#### Solution

\begin{align}
AX &= B \\
A^{-1} AX &= A^{-1} B \\
I_2 X &= A^{-1} B \\
X &= \begin{bmatrix} -14 + \frac{21}{2} \\ 7 + \frac{-7}{2} \end{bmatrix} = \begin{bmatrix} \frac{-7}{2} \\ \frac{7}{2} \end{bmatrix}
\end{align}

### Question

(ex. 9, p. 124, sec. 2.3, Kolman) Is the following matrix singular?

$$
\begin{bmatrix}
1 & 3 \\
2 & 6
\end{bmatrix}
$$

#### Solution

Yes it is singular, i.e non-invertible, Since its determinant is
$$
1 \cdot 6 - 3 \cdot 2 = 6 - 6 = 0
$$

### Question

(ex. 2 (c), p. 155, sec. 3.2, Kolman) Compute the determinant using properties of determinants.

$$
A =
\begin{bmatrix}
3 & 4 & 2 \\
2 & 5 & 0 \\
3 & 0 & 0
\end{bmatrix}
$$

#### Solution

Consider the matrix obtained by interchanging first and third rows

$$
A_{R_1 \leftrightarrow R_3} =
\begin{bmatrix}
3 & 0 & 0 \\
2 & 5 & 0 \\
3 & 4 & 2
\end{bmatrix}
$$

From theorem 3.7 (page 149), Its determinant is the multiplication of diagonal entries.

$$
2 \cdot 5 \cdot 3 = 30
$$

From theorem 3.2 (page 147), $\det A = - \det A_{R_1 \leftrightarrow R_3} = -30$.

### Question

(ex. 7 (b), p. 169, sec. 3.4, Kolman)  Find the inverse if it exists using the adjoint matrix method.

\begin{align}
A =
\begin{bmatrix}
4 & 2 & 2 \\
0 & 1 & 2 \\
1 & 0 & 3
\end{bmatrix}
\end{align}

#### Solution

Minors
\begin{align}
\det M_{11} &= \begin{vmatrix} 1 & 2 \\ 0 & 3 \end{vmatrix} = (3) - (0) = 3 \\
\det M_{12} &= \begin{vmatrix} 0 & 2 \\ 1 & 3 \end{vmatrix} = (0) - (2) = -2 \\
\det M_{13} &= \begin{vmatrix} 0 & 1 \\ 1 & 0 \end{vmatrix} = (0) - (1) = -1 \\
\det M_{21} &= \begin{vmatrix} 2 & 2 \\ 0 & 3 \end{vmatrix} = (6) - (0) = 6 \\
\det M_{22} &= \begin{vmatrix} 4 & 2 \\ 1 & 3 \end{vmatrix} = (12) - (2) = 10 \\
\det M_{23} &= \begin{vmatrix} 4 & 2 \\ 1 & 0 \end{vmatrix} = (0) - (2) = -2 \\
\det M_{31} &= \begin{vmatrix} 2 & 2 \\ 1 & 2 \end{vmatrix} = (4) - (2) = 2 \\
\det M_{32} &= \begin{vmatrix} 4 & 2 \\ 0 & 2 \end{vmatrix} = (8) - (0) = 8 \\
\det M_{33} &= \begin{vmatrix} 4 & 2 \\ 0 & 1 \end{vmatrix} = (4) - (0) = 4 \\
\end{align}

Cofactors
\begin{align}
A_{11} &= (-1)^{1+1} \det M_{11} = 3 \\
A_{12} &= (-1)^{1+2} \det M_{12} = 2 \\
A_{13} &= (-1)^{1+3} \det M_{13} = -1 \\
A_{21} &= (-1)^{2+1} \det M_{21} = -6 \\
A_{22} &= (-1)^{2+2} \det M_{22} = 10 \\
A_{23} &= (-1)^{2+3} \det M_{23} = 2 \\
A_{31} &= (-1)^{3+1} \det M_{31} = 2 \\
A_{32} &= (-1)^{3+2} \det M_{32} = -8 \\
A_{33} &= (-1)^{3+3} \det M_{33} = 4 \\
\end{align}

Determinant (first column)
$$
a_{11} A_{11} + a_{21} A_{21} + a_{31} A_{31} = 4 (3) + 0 + 1 (2) = 12 + 2 = 14
$$

Adjoint Matrix

$$
adj \, M =
\begin{bmatrix}
3 & -6 & 2 \\
2 & 10 & -8 \\
-1 & 2 & 4
\end{bmatrix}
$$

Inverse Matrix
$$
M^{-1} = \frac{1}{\det M} \cdot adj \, M =
\begin{bmatrix}
\frac{3}{14} & \frac{-3}{7} & \frac{1}{7} \\
\frac{1}{7} & \frac{5}{7} & \frac{-4}{7} \\
\frac{-1}{14} & \frac{1}{7} & \frac{2}{7}
\end{bmatrix}
$$
