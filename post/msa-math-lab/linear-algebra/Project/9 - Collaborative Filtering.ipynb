# Collaborative Filtering

```python
import numpy as np
```

```python
# watching record matrix
# user x TV-shows
M = np.array([
    [30, 35, 0, 0],
    [5, 0, 0, 30],
    [0, 0, 32, 20],
    [0, 0, 34, 5]
])
```

```python
users = np.array([ "Antar", "Tameem", "Logain", "Jumana" ])
tvShows = np.array([ "Game Of Thrones", "Peaky Blenders", "Hannah Montana", "The IT Crowded" ])
```

Tv-shows seen by Logain `M[2]` but not Tameem `M[1]`.

```python
# tv-shows watched by Logain
M[2] > 0
```

```python
# tv-shows not watched by Tameem
M[1] == 0
```

```python
# tv-shows watched by Logain but not Tameem
(M[2] > 0) & (M[1] == 0)
```

```python
# tv-shows names
tvShows[ (M[2] > 0) & (M[1] == 0) ]
```

Average Episodes Watched

```python
# input. vector u
#        vector v
# output. vector u+v
def addVec(u, v):
  return u + v
```

```python
from functools import reduce
```

```python
# add all vectors
reduce( addVec, M )
```

```python
# add vectors and divide by their size
reduce( addVec, M ) / len(M)
```

Users who watched episodes more than Jumana

```python
# boolean mask
boolMask = np.array([
  # if episodes sum of user i is greater than Jumana's
  # True is added. Otherwise False is added.
  True if np.sum(i) > np.sum(M[3]) else False
    # loop on each user i (rows of the matrix)
    for i in M
])
boolMask
```

```python
users[ boolMask ]
```
