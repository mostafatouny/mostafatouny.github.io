**List Comprehension**

```python
l = [2, 4, 6, 8]
l[2]
```

```python
l = [1, 3, 5, 7]
l
```

```python
for i in range(4):
  print(i)
```

```python
l = [ i for i in range(5) ]
```

```python
l
```

```python
l = [ 2*i for i in range(5) ]
l
```

```python
ev = [ i for i in range(10) if (i%2 == 0) ]
```

```python
ev
```

```python
od = [ i for i in range(10) if not (i%2 == 0) ]
```

```python
od
```

```python
[ (2*i) + 1 for i in range(5) ]
```

```python
[ od[i]+ev[i] for i in range(5) ]
```

**2d List & Matrix Addition**

```python
A = [
    [1, 3],
    [2, 4]
  ]
```

```python
B = [
    [2, 6],
    [10, 8]
  ]
```

```python
B[1][0]
```

```python
[ A[0][j] + B[0][j] for j in range(2) ]
```

```python
[ A[1][j]+B[1][j] for j in range(2) ]
```

```python
[
    [ A[i][j]+B[i][j] for j in range(2) ]
      for i in range(2)
]
```

```python
A
```
