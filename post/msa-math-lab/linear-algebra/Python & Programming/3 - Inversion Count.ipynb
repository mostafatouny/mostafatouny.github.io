```python
# input: list a
# output: multiplication of its elements
def lisMul(a):
  res = 1
  for i in a:
    res *= i
  return res
```

```python
# input: list a
# output: number of inversions, i.e a[i] > a[j] where i < j
def inversionCount(a):
  # number of inversions
  count = 0

  # for each index i of list a
  for i in range(len(a)):
    # j = i, i+1, ..
    for j in range(i, len(a)):
      # if preceeding and greater
      if a[i] > a[j]:
        # count
        count += 1

  return count

# input: list a
# output: 1 if it has even inversions
#         -1 if it has odd inversions
def parity(a):
  return 1 if inversionCount(a) % 2 == 0 else -1
```

```python
#####
## Optional
#####

# input: list a
# output: all permutations of elements of a
def permutations(a):
  # base case
  if len(a) == 1:
    return [ [ a[0] ] ]

  # smaller instance
  smaller_instance = permutations(a[1:])

  # main instance
  main_instance = [
    # add a[0] to ith position, shifting substring to right
    perm[0:i] + [ a[0] ] + perm[i:]

    # for each permutation
    for perm in smaller_instance
      # for each position i = 0,1, .., n; i = n appends to last
      for i in range( len(perm)+1 )
  ]

  return main_instance
```

```python
#####
## Optional
#####

# input: matrix M
# output: determinant of M
def det_nxnMat(M):
  # dimension of the matrix
  n = len(M)

  # generate all permutations
  perms = permutations( [ i for i in range(n) ] )

  sum = 0

  # for each permutation
  for perm in perms:
    # elements of the matrix, for each row, induced by the permutation
    tmp = [ M[i][perm[i]] for i in range(n) ]
    # multiply, then to sum plus if even parity, and subtract if odd parity
    sum += parity(perm) * lisMul(tmp)

  return sum
```

```python
# kolman, p. 409
det_nxnMat([
    [1,2,3],
    [2,1,3],
    [3,1,2]
])
```

```python
# input. list l
# output. count of inversions number
def invCou(l):
  cou = 0
  # i = 0, 1, 2
  for i in range(0, len(l) ):
    for j in range(i+1, len(l)):
      if l[j] < l[i]:
        cou = cou + 1
  return cou
```
