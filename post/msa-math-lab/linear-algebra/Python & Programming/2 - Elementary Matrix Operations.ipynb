Row scalar multiplication

```python
# scalar multiplication on ith row
# Input. Matrix M
#        Row position i
#        Scalar c
# Output. None. It modifies matrix in place
def scalMul(M, i, c):
  n = len(M)
  for j in range(n):
    # row i is fixed
    M[i][j] = c * M[i][j]
```

```python
scalMul(M, 0, 2)
```

```python
# Interchanging rows i and j
# Input. Matrix M
#        Row position i
#        Row position q
# Output. None. It modifies matrix in place
def swapRowsMat(M, i, q):
  n = len(M)
  for j in range(n):
    # row i and q are fixed
    M[i][j], M[q][j] = M[q][j], M[i][j]
```

```python
M
```

```python
# swap first and second rows
swapRowsMat(M, 0, 1)
```

Row-reduce an upper-triangular matrix using recursion.

Example.

\begin{align}
\left [
\begin{array}{ccc|c}
        1 & 1 & 4 & 32 \\
        0 & 1 & 1 & 12 \\
        0 & 0 & 3 & 10  \\
    \end{array}
\right ]
\rightarrow
\left [
\begin{array}{ccc|c}
        1 & 0 & 0 & 10 \\
        0 & 1 & 0 & \frac{26}{3} \\
        0 & 0 & 1 & \frac{10}{3}  \\
    \end{array}
\right ]
\end{align}

```python
#####
## Optional
#####

# input: nxn Matrix M
#        Row Index r
def rowReduce(M, r):
  # if all rows are updated, we are done
  if r == -1:
    return M

  # dimension
  n = len(M)

  # update row in position r by its element in position r (diagonal entries)
  M[r] = M[r] / M[r][r]

  # update upper rows
  for i in range(r-1, -1, -1):
    M[i] = M[i] + ( M[r] * (-M[i][r]) )

  # recursively call smaller instance
  return rowReduce(M, r-1)
```
