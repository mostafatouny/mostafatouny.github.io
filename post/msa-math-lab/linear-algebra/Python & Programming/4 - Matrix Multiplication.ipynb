**Matrix Multiplication**

```python
M = [
    [1, 2],
    [3, 4]
]

N = [
    [5, 6],
    [7, 8]
]
```

```python
# Compute MN[0][0]
sum = 0
# 2 is the dimension of both matrices
for i in range(2):
    # row of M and column of N are fixed
    sum = sum + (M[0][i] * N[i][0])
sum
```

```python
# Input. Matrix M
#        Matrix N
#        row position i
#        column position j
# Output. MN[i][j]
def matCell(M, N, i, j):
  # dimension n
  n = len(M)
  sum = 0
  for z in range(n):
    # row of M and column of N are fixed
    sum = sum + (M[i][z] * N[z][j])
  return sum
```

```python
matCell(M, N, 0, 0)
```

```python
# input. nxn Matrix M
#        nxn Matrix N
def matMul(M, N):
  n = len(M)

  O = [
      [ matCell(M, N, j, i) for i in range(n) ]
        for j in range(n)
  ]

  return O
```

```python
matMul(M, N)
```
