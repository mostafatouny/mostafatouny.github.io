([Beginning Algebra](https://math.libretexts.org/Bookshelves/Algebra/Beginning_Algebra/04%3A_Solving_Linear_Systems/4.04%3A_Applications_of_Linear_Systems), 4.4.3) 32 bills consist of 5\$ and 10$ values, with a total value of 220\$. How many bills of each value do we have?

**Formulation.**

System of Equations

\begin{align}
x + y &= 32 \\
5x + 10y &= 220
\end{align}


Matrix

\begin{align}
\begin{bmatrix}
1 & 1
\\ 5 & 10
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
32 \\
220
\end{bmatrix}
\end{align}

Augmented Matrix

\begin{align}
\left [
\begin{array}{cc|c}
        1 & 1 & 32 \\
        5 & 10 & 220 \\
    \end{array}
\right ]
\end{align}


```python
# construct matrix M
M = np.array([
  [1, 1],
  [5, 10]
])
```

```python
# construct vector b
b = np.array([
    32, 220
])
```

```python
# solve the system, i.e finding vector x
v = np.linalg.solve(M, b)
v
```

So $x = 20$ and $y = 12$. Let's verify the answer.

```python
np.dot(M, v)
```

Coinciding with given $b$.
