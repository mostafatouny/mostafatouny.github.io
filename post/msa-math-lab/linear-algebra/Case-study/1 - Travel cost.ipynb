**Story** (Rosen, p.745, figure 3). You are given a map of gates fees between some states of the US. You want to compute the total income the gate of _Chicago_ gets for each city, for the day where:

| City | Cars Passed |
|----- | ------ |
| Boston | 100 |
| New York | 30 |
| Dallas | 220 |
| Los Angeles | 77 |
| San Francisco | 160 |

**Formulation**

Cities-Indices Mapping

| Index | City |
|-----|-------|
| 0 | San Francisco |
| 1 | Los Angeles |
| 2 | Denver |
| 3 | Dallas |
| 4 | Chicago |
| 5 | Boston |
| 6 | New York |

Distance Matrix. $M[i,j]$ is cost from city i to city j by a single road, and zero if no road exists.

\begin{align}
M =
\begin{bmatrix}
0 & 400 & 1000 & 0 & 1500 & 0 & 0 \\
400 & 0 & 0 & 1100 & 1400 & 0 & 0 \\
1000 & 0 & 0 & 600 & 500 & 0 & 0 \\
0 & 1100 & 600 & 0 & 600 & 0 & 1200 \\
1500 & 1400 & 500 & 600 & 0 & 900 & 700 \\
0 & 0 & 0 & 0 & 900 & 0 & 300 \\
0 & 0 & 0 & 1200 & 700 & 300 & 0
\end{bmatrix}
\end{align}

Cars passed.

\begin{align}
v =
\begin{bmatrix}
160 \\
77 \\
0 \\
220 \\
0 \\
100 \\
30 \\
\end{bmatrix}
\end{align}

Now $(Mv)[4]$ is the total income for Chicago, for each city.

```python
M = np.array([
  [0, 400, 1000, 0, 1500, 0, 0],
  [400, 0, 0, 1100, 1400, 0, 0],
  [1000, 0, 0, 600, 500, 0, 0],
  [0, 1100, 600, 0, 600, 0, 1200],
  [1500, 1400, 500, 600, 0, 900, 700],
  [0, 0, 0, 0, 900, 0, 300],
  [0, 0, 0, 1200, 700, 300, 0]
])
```

```python
v = np.array([
    160, 77, 0, 220, 0, 100, 30
])
```

```python
(M * v)[4]
```

Answer is

| City | Income |
| ---- | ----- |
| San Francisco | 240 000 |
| Los Angeles | 107 800 |
| Dallas | 132 000 |
| Boston | 90 000 |
| New York | 21 000 |
