## Google Colab

[Google Colab: Everything You Need to Know](https://geekflare.com/google-colab/)

**Cloud**
- Your code runs on other computers.
- Avoid the hassle of hardware's CPU, GPU, Memory, TPUs, ..etc.
- Avoid the hassle of setting-up a local environment.

**Google's Features**
- FREE
- Notebook sharing.
- Pre-installed libraries.
- Cloud-storage.

**Create a Notebook**

- Google Chrome browser is recommended.
- Log-in to your Google's account (Use MSA's account).
- `New -> More -> Google Colaboratory`.

Screenshot_2024-02-08_08-42-30.png

**Cloning a Notebook**

- `File -> Save a copy in Drive`

Screenshot_2024-02-06_12-29-28.png

**Downloading a Notebook**
- `File -> Download -> Download as .ipynb`.

**Session**

Disconnect (You may use paid cloud service some day)
- `Runtime -> Disconnect` before closing the tab.

Restart (Get rid of all cached variables)
- `Runtime -> Restart Session`

Screenshot_2024-02-08_08-38-52.png

**Tips**
- Execute all cells: `Runtime -> Run All`.
- run selected cell: `Shift+Enter`.
-  Command Palette: `CTRL+SHIFT+P`.
