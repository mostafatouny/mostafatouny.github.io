**Riemann Sum by right Hand**

```python
# ith right hand x point
def c_i(a, delta, i):
  return a + (i * delta)

# riemann sum by right hand
# function f
# n sub-intervals
# interval [a, b]
def riemannSum_rightHand( f, n, a, b ):
  delta = (b - a) / n
  sum = 0

  # i = 1, 2, .., n
  for i in range(1, n+1):
    # right hand x point
    rightHand = c_i(a, delta, i)
    # function value at right hand, multiplied by delta length
    sum += delta * f( rightHand )

  return sum

# a list of riemann sums for n = 1, 2, .., k
def listOf_riemannSum_rightHand(f, k, a, b):
  return [ riemannSum_rightHand(f, n, 0, 1) for n in range(1, k+1)  ]
```

**Computing values for function $f$ solved in _On-Paper_**

```python
# f(x) = 1 - x^2
def f(x):
  return 1 - x**2
```

```python
# riemann sum by right hand
# for function f
# on interval [0, 1]
riemannSum_rightHand(f, 1, 0, 1)
```

```python
# riemann sum for n = 1, 2, .., 10
listOf_riemannSum_rightHand(f, 10, 0, 1)
```

Convergence to $\frac{2}{3}$ is observed.
