### Question 1

(ex. 15, sec. 7.5, p. 540, Briggs) Evaluate

$$
A = \int \frac{6}{x^2 - 1} \, dx
$$

#### Solution

\begin{align}
\frac{6}{x^2 - 1} &= \frac{6}{(x+1)(x-1)} \\
&= \frac{A}{x+1} + \frac{B}{x-1} \\
6 &= (x-1)A + (x+1)B \\
0x + 6 &= x(A+B) + (B-A)
\end{align}
It follows
\begin{align}
0 &= A+B \\
6 &= B-A \\
A &= -3 \\
B &= 3
\end{align}
So
\begin{align}
A &= \int \frac{-3}{x+1} + \frac{3}{x-1} \\
&= -3 \int \frac{1}{x+1} + 3 \int \frac{1}{x-1} \\
&= -3 \ln |x+1| + 3 \ln |x-1|
\end{align}

### Question 2

(ex. 30, sec. 7.5, p. 540, Briggs) Evaluate

$$
A = \int \frac{1}{x^3 - 2x^2 - 4x + 8} \, dx
$$

#### Solution

\begin{align}
x^3 - 2x^2 - 4x + 8 &= x^3-4x - 2x^2 + 8 \\
&= x(x^2 - 4) - 2(x^2 - 4) \\
&= (x^2-4) (x-2) \\
&= (x+2) (x-2)^2
\end{align}

Then
\begin{align}
\frac{1}{x^3 - 2x^2 - 4x + 8} &= \frac{1}{(x+2)(x-2)^2} = \frac{A}{(x+2)} + \frac{B}{(x-2)} + \frac{C}{(x-2)^2} \\
1 &= (x-2)^2 A + (x+2)(x-2) B + (x+2) C \\
0x^2 + 0x + 1 &= (A + B)x^2 + (C-4A)x + (4A-4B+2C)
\end{align}

It follows
\begin{align}
0 &= A+B \\
0 &= C-4A \\
1 &= 4A-4B+2C
\end{align}

In turn
\begin{align}
B = -A \\
A = \frac{C}{4}
\end{align}

Concluding
$$
1 = C + C + 2C = 4C
$$

So
\begin{align}
A &= \frac{1}{16} \\
B &= \frac{-1}{16} \\
C &= \frac{1}{4}
\end{align}

Finally
\begin{align}
A &= \frac{1}{16} \int \frac{1}{x+2} + \frac{-1}{16} \int \frac{1}{x-2} + \frac{1}{4} \int \frac{1}{(x-2)^2} \\
&= \frac{1}{16} \ln |x+2| + \frac{-1}{16} \ln |x-2| + \frac{-1}{4} \frac{1}{x-2} + C
\end{align}

### Question 3

(ex. 48, sec. 8.4, page. 541, Briggs) Evaluate
$$
A = \int \frac{2x + 1}{x^2 + 4} \, dx
$$

#### Hints

- You will use trigonometric substitution.

#### Solution

$x^2 + 4$ is irreducible. Set

\begin{align}
\frac{2x + 1 }{x^2 + 4} &= \frac{Ax + B}{x^2 + 4} \\
2x + 1 &= Ax + B \\
2 &= A  \\
1 &= B
\end{align}
It follows
\begin{align}
A &= \int \frac{2x}{x^2+4} \, dx + \int \frac{1}{x^2+4} \, dx \\
&= \int \frac{du}{u} + \int \frac{1}{(2 \tan \theta)^2 + 2^2} \, 2 \sec^2 \theta \\
&= \ln |x^2 + 4| + \int \frac{1}{2^2 \sec^2 \theta} 2 \sec^2 \theta \\
&= \ln |x^2 + 4| + \frac{1}{2} \int 1 \, d\theta \\
&= \ln |x^2 + 4| + \frac{1}{2} \cdot \theta \\
&= \ln |x^2 + 4| + \frac{1}{2} \tan^{-1}(\frac{x}{2})
\end{align}
