### Question

(ex. 45, sec. 5.5, p. 383, Briggs) Evaluate

$$
A = \int_{\pi/4}^{\pi/2} \frac{\cos x}{\sin^2 x} \, dx
$$

#### Solution

Set
\begin{align}
u &= \sin x \\
du &= \cos x \, dx
\end{align}
Note
\begin{align}
u(\frac{\pi}{2}) &= \sin(\frac{\pi}{2}) = 1 \\
u(\frac{\pi}{4}) &= \sin(\frac{\pi}{4}) = \frac{\sqrt{2}}{2}
\end{align}
Then
\begin{align}
A &= \int_{ \frac{\sqrt{2}}{2} }^1 \frac{1}{u^2} \, du \\
&= -1 \cdot u^{-1} \bigg |_{\frac{\sqrt{2}}{2}}^{1} \\
&= -1 \cdot (1)^{-1} - \left (- \left (\frac{\sqrt{2}}{2} \right )^{-1} \right ) \\
&= -1 + \frac{2}{\sqrt{2}} \\
&= \sqrt{2} - 1
\end{align}

### Question

(ex. 52, sec. 5.5, p. 383, Briggs) Evaluate
\begin{align}
A = \int_0^{\ln 4} \frac{e^x}{3 + 2e^x} dx
\end{align}

#### Hint

- (Abstract; Pattern) Decompose the algebraic expressions. Imagine one of them the derivative of the other.
- Multiply by $\frac{2}{2}$.
- Recall $[e^x]' = e^x$.

#### Solution

\begin{align}
A &= \frac{1}{2} \int_0^{\ln 4} \frac{2e^x}{3 + 2e^x} dx \\
g(x) &= 3 + 2e^x \\
g'(x) &= 2e^x \\
f(x) &= \frac{1}{x} \\
A &= \frac{1}{2} \int_0^{\ln 4} f(g(x)) \cdot g'(x) \\
&= \frac{1}{2} \int_{g(0)}^{g(\ln 4)} f(u) \, du \\
&= \frac{1}{2} \int_5^{11} \frac{1}{u} \, du \\
&= \frac{1}{2} \cdot \ln u \bigg |_5^{11} \\
&= \frac{1}{2} \cdot \ln 11 - \ln 5 \\
&= \frac{\ln(11/5)}{2}
\end{align}
