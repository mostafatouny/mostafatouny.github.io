### Question

(ex. 23, p. 581, sec. 8.1, Briggs) Solve for $y$ given

$$
y'(t) = 1 + e^t \quad y(0) = 4
$$

#### Solution

\begin{align}
y(t) &= \int 1 + e^t = t + e^t + C \\
y(0) &= 0 + e^0 + C = 4 \\
&= 1 + C = 4 \\
C &= 3 \\
y(t) &= t + e^t + 3
\end{align}

### Question

(ex. 24, p. 595, sec. 8.1, Briggs) Solve

$$
\sec x \, y'(x) = y^3
$$

#### Solution

\begin{align}
\frac{1}{y^3} \cdot y'(x) &= \frac{1}{\sec x} \\
\int y^{-3} y' \, dx &= \int \cos x \, dx \\
\int y^{-3} \, dy &= \sin x + c \\
\frac{-1}{2} y^{-2} &= \sin x + c \\
\frac{-1}{2} \cdot y^{-2} &= \\
\frac{1}{y^2} &= -2 \sin x + c
\end{align}

Since $y(0) = 3$

\begin{align}
\frac{1}{3^2} &= -2 \sin 0 + c \\
\frac{1}{9} &= c
\end{align}

Therefore

$$
\frac{1}{y^2} = -2 \sin x + \frac{1}{9}
$$

### Question

(ex. 11, p. 603, sec. 8.4, Briggs) Solve the first-order differential equation

$$
y'(t) = 3y - 6, y(0) = 9
$$

#### Solution

\begin{align}
b &= -6 \\
k &= 3 \\
y(t) &= C e^{3t} + \frac{6}{3} \\
y(0) &= C + 2 = 9 \\
C &= 7 \\
y &= 7e^{3t} + 2
\end{align}
