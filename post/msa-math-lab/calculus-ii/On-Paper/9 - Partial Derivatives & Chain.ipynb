### Question

(ex. 7, p. 925, sec 13.4, Briggs) Evaluate the partial derivatives of

$$
f(x,y) = 3x^2 + 4y^3
$$

#### Solution

\begin{align}
f_x(x,y) &= 6x + 0 = 6x \\
f_y(x,y) &= 0 + 12y^2 = 12y^2
\end{align}

### Question

(ex. 34, p. 926, sec 13.4, Briggs) Evaluate the partial derivatives of

$$
H(x,y) = \sqrt{4 + x^2 + y^2}
$$

#### Solution

\begin{align}
H_x(x,y) &= \frac{1}{2} (x^2 + y^2 + 4)^{-1/2} \cdot (2x + 0) \\
&= \frac{x}{\sqrt{x^2 + y^2 + 4}}
\end{align}

$H_y$ is similar.
