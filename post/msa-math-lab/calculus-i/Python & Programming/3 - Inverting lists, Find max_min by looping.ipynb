### Finding Max/Min by Naive Looping

```python
def findMax_naive(f, a, b):
  x = np.arange(a, b)
  y = f(x)

  max = 0
  for v in y:
    if v > max:
      max = v

  return max
```

```python
def f(x):
  return x

findMax_naive(f, -20, 20)
```
