## Case-study: Maximizing Car Renting Revenue

(Ex. 4.35, [Sec. 4.7](https://openstax.org/books/calculus-volume-1/pages/4-7-applied-optimization-problems), Calculus I) Owners of a car rental company have determined that if they charge customers $p$ dollars per day to rent a car, where $50 \leq p \leq 200$, Then 1000 cars will be rented but each increment in price loses 5 cars. How much should they charge to maximize their revenue?


### Hint

- Formulation is a necessary step to use calculus toolboxes.

### Solution

**Formulation.** Number of cars to be rented is $c(p) = 1000 - 5(p - 50)$. Revenue is $r(p) = p \cdot c(p) = 1000p - 5p^2 + 250p = 1250p - 5p^2$.

```python
def cars(p):
  return 1000 - (5*(p-50))

def revenue(p):
  return p * cars(p)
```

## Case-Study: Predicting Which Air Line will be More Profitable?

You are given an offer to work in two international air lines companies: _A_ and _B_. You want to decide which of them is more profitable on both the short-term and long-term.

Answer by plotting and interpreting the tangent line.

```python
# Company A

def profit(days):
  return 3 * days

plotWithTangent(profit, 1, 0, 10)
```

```python
# Company B

def profit(days):
  return 2 * (days**2)

plotWithTangent(profit, 2, 0, 10)
```
