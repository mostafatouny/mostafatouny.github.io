### Question

(ex. 24, p. 148, sec. 3.2, Briggs)  Find the derivative of
$$
s(t) = 4 \sqrt{t} - \frac{1}{4} t^4 + t + 1
$$

#### Solution

\begin{align}
s'(t) &= 4 \cdot [t^{1/2}]' - \frac{1}{4} [t^4]' + 1 + 0 \\
&=  4 \cdot \frac{1}{2} \cdot t^{-1/2} - \frac{1}{4} \cdot 4 \cdot t^3 + 1 \\
&= \frac{2}{t^{1/2}} - t^3 + 1
\end{align}

### Question

(ex. 27, p. 149, sec. 3.2, Briggs) Differentiate
$$
h(x) = (x^2 + 1)^2
$$
Using both the product rule and chain rule.

#### Solution

By product rule,
\begin{align}
h'(x) &= (x^2 + 1) \cdot [x^2 + 1]' + [x^2 + 1]' \cdot (x^2 + 1) \\
&= 2 \cdot (x^2 + 1) \cdot 2x \\
&= 4x^3 + 4x
\end{align}

By chain rule,

Let $f(x) = x^2 + 1$ and $g(x) = x^2$. Then $h(x) = g(f(x))$. It follows
\begin{align}
g'(x) &= 2x \\
f'(x) &= 2x \\
h'(x) &= g'(f(x)) \cdot f'(x) \\
&= 2 (x^2 + 1) \cdot 2x \\
&= 4x^3 + 4x
\end{align}

### Question

(ex. 48, p. 149, sec. 3.2, Briggs) Differentiate
$$
f(x) = 10e^x
$$

#### Solution

\begin{align}
f'(x) &= 10 \cdot [e^x]' \\
&= 10e^x
\end{align}

### Question

(ex. 33, p. 167, sec. 3.4, Briggs) Differentiate
$$
y = \sec x + \csc x
$$

#### Solution

By _theorem 3.13_ (p. 165),
$$
y' = \sec x \tan x + - \csc x \cot x
$$

### Question

(ex. 55, p. 207, sec. 3.8, Briggs) Differentiate
$$
y = 4 \log_3 (x^2 - 1)
$$

#### Solution

By _theorem 3.20_, p. 205, and the chain rule,

\begin{align}
y' &= 4 \cdot [\log_3 (x^2 - 1)]' \\
&= 4 \cdot \frac{1}{ (x^2 - 1) \cdot \ln 3 } \cdot [x^2 - 1]' \\
&= \frac{8x}{ (x^2 - 1) \cdot \ln 3 }
\end{align}
