### Question

(example 3.1, page 5, sheet) Evaluate

$$
\lim_{x \rightarrow \infty} x \sin \frac{1}{x}
$$

#### Solution

\begin{align}
\lim_{x \rightarrow \infty} x \sin \frac{1}{x}
&= \lim_{x \rightarrow \infty} \frac{\sin(1/x)}{1/x} \\
&= \frac{0}{0} \quad \text{indeterminate}
\end{align}
Apply L'Hopital,
\begin{align}
\lim_{x \rightarrow \infty} \frac{\sin(1/x)}{1/x} &= \lim_{x \rightarrow \infty}
 \frac{\frac{-1}{x^2} \cos(\frac{1}{x})}{\frac{-1}{x^2}} \\
&= \lim_{x \rightarrow \infty} \cos( \frac{1}{x} ) = 1
\end{align}

### Question

(example 3.2, page 6, sheet) Evaluate

$$
\lim_{x \rightarrow 0} \frac{1}{\sin x} - \frac{1}{x}
$$

#### Solution

\begin{align}
\lim_{x \rightarrow 0} \frac{1}{\sin x} - \frac{1}{x} &= \lim_{x \rightarrow 0} \frac{x - \sin x}{x \sin x} \\
&= \lim_{x \rightarrow 0} \frac{1 - \cos x}{\sin x + x \cos x} \\
&= \lim_{x \rightarrow 0} \frac{\sin x}{2 \cos x - x \sin x} \\
&= 0
\end{align}


### Question

(example 3.3, page 6, sheet) Evaluate

$$
\lim_{x \rightarrow \infty} x^{1/x}
$$

#### Hints

- Recall $e^{\ln x} = x$.
- Note $\ln x$ is defined only for $x > 0$.

#### Solution

\begin{align}
\lim_{x \rightarrow \infty} x^{1/x} &= \lim_{x \rightarrow \infty} \exp( \ln x^\frac{1}{x} ) \\
&= \lim_{x \rightarrow \infty} \exp(\frac{\ln x}{x} ) \\
&= \lim_{x \rightarrow \infty} \exp(\frac{1/x}{1} ) \\
&= 1 \quad \text{as } e^0 = 1
\end{align}
