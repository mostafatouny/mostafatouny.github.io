### Question

(ex. 14, p. 10, sec. 1.1, Briggs) Decide the domain and range of
\begin{align}
g(y) = \frac{y+1}{(y+2)(y-3)}
\end{align}

#### Hints

- (draw a diagram) Graph the function.
- (observation) Note it is a rational function of polynomials.
- (proof technique) Set $g(y) = z$ for arbitrary $z$.

#### Solution

The domain is $(-\infty , -2) \cup (-2,3) \cup (3,\infty)$. The range is $\mathcal{R}$, All real numbers.

### Question

(ex. 84, p. 11, sec. 1.1, Briggs)

**a.** if $f$ is even, $f(0)$ is defined, is it necessarily $f(0) = 0$?

**b.** if $f$ is odd, $f(0)$ is defined, is it necessarily $f(0) = 0$?

#### Hints

- Just apply the definition
- What can you conclude if $x = -x$?

#### Solution

**a.** No. Counter-example: $f(x) = x^2 + 3$ is even but $f(0) \neq 0$.

**b.** Yes. By definition $f(0) = f(-0) = -f(0)$ concluding $f(0) = 0$.

### Question

([6.7.4](https://math.libretexts.org/Courses/Highline_College/Math_081_091%3A_CAM_Aligned_Textbook/06%3A_Linear_Equations_and_Functions/6.07%3A_Find_the_Equation_of_a_Line), highline college) Find the line equation which passes through points $(0,−4)$ and $(3,−2)$.

#### Hints

- (pattern) Recall the fact the slope is constant among any two points of a line.
- (Observation) Note $f(0) = m \cdot 0 + b = b$.

#### Solution

Slope $m = \frac{-2 - (-4)}{3 - 0} = \frac{2}{3}$

Intercept $b = f(0) = -4$.

Equation is $y = \frac{2}{3}x - 4$.
