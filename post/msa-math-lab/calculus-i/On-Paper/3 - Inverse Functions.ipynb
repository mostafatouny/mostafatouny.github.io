### Question

(Ex. 27, p. 36, Sec 1.1, Briggs) Find the inverse of

$$
f(x) = \sqrt{x + 2} \quad \text{for } x \geq -2
$$

#### Solution

\begin{align}
x &= \sqrt{y+2} \\
x^2 &= y + 2 \\
x^2 - 2 &= y \\
f^{-1}(x) &= x^2 - 2
\end{align}

##### Remark

$$
f(x) = 2^x \\
f^{-1}(x) = \log_2(x) \\
f(3) = 8 \\
f^{-1}(8) = 3 \\
f^{-1}(f(x)) = x \\
\log_2(2^3) = 3
$$

### Question

(Ex. 28, p. 36, Sec 1.3, Briggs) Find the inverse of

\begin{align}
f(x) = \frac{2}{x^2 + 1} \quad \text{for } x \geq 0
\end{align}


#### Solution

Switching $x$ and $y$, $x = \frac{2}{y^2 + 1}$. It follows $|y| = \sqrt{\frac{2}{x} - 1}$. Since domain of $f$ is $[0, \infty)$, It is also the range of $f^{-1}$. So $|y| = y$.

\begin{align}
x &= \frac{2}{y^2 + 1} \\
2 &= x(y^2 + 1) \\
2 &= xy^2 + x \\
2 - x &= xy^2 \\
\frac{2}{x} - 1 &= y^2 \\
\sqrt{ \frac{2}{x} - 1 } &= \sqrt{y^2} \\
&= \pm y \\
&= y
\end{align}

$$
y = 3 \\
|y| = y = 3 \\
y = -3 \\
|y| = -y = -(-3) = 3 \\
|y| \neq y
$$
