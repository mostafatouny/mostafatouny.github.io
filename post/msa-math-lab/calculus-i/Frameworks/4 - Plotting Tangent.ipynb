## Helper Functions

```python
# @title
# Derivative
def deriv(f, x):
  # step-size
  h = 0.000000001
  # definition
  return (f(x+h) - f(x)) / h

# Plot with tangent
def plotWithTangent(f, x_0, a, b):
  # x sequence
  x = np.linspace(a, b, 200)
  # y sequence
  y = f(x)
  # y_0
  y_0 = f(x_0)
  # tangent line
  y_tan = deriv(f, x_0) * (x - x_0) + y_0

  # graph
  plt.plot(x, y, 'b-')
  # tangent line
  plt.plot(x, y_tan, 'r', linestyle='--')
  plt.axis([a, b, a, b])
  # x axis line
  plt.axhline(0, color='black', linewidth=0.5) # x = 0
  # y axis line
  plt.axvline(0, color='black', linewidth=0.5) # y = 0
  # grid
  plt.grid()
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(f'Plot of a function with tangent line at x={x_0}')
  plt.show()
```

## Main

**Geogebra.** Tangent line

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Plotting a function with tangent line**

```python
# f(x) = x^2
def f(x):
  return x**2

# input
## function
## point slope at
## starting point of the interval
## ending point of the interval
plotWithTangent(f, 1, -5, 5)
```

```python
# f(x) = e^x
def f(x):
  return np.exp(x)

plotWithTangent(f, 0.5, -1, 3)
```
