Quantification

All students passed the exam $\leftrightarrow$ No student did not pass the exam.

For each student $i$, $i$ had passed the exam $\leftrightarrow$ It is not the case, That there exists a student who did not pass the exam.

S is subset of Q $\leftrightarrow$ For each element $i$ of $S$, $i$ is also an element of $Q$ $\leftrightarrow$ It is not the case, There is an element $i$ in $S$ not in $Q$.

```python
# input: sets S and Q
# output: True iff S is a subset of Q
def isSubset(S, Q):
  # for each element of S
  for i in S:
    # if there is a single element in S, not in Q
    if not i in Q:
      # then conclude S isn't a subset of Q
      return False
    # end here
  # ends here

  # we reach here only if
  # no element of S, is not in Q
  return True
```

\begin{align}
  isSubset(S,Q) = False &\leftrightarrow \exists i \in S, i \not\in Q \\
  isSubset(S,Q) = True  &\leftrightarrow \forall i \in S, \neg (i \not\in Q) \\
                        &\leftrightarrow \forall i \in S, i \in Q
\end{align}

```python
isSubset({}, {0, 2, 5})
```

```python
def isSubset_bad(S, Q):
  # count elements of S contained in Q
  counter = 0
  # try all elements in S
  for i in S:
    # if contained, add to counter
    if i in Q:
      counter = counter + 1
  # if count of contained elements is the same as length of S
  if counter == len(S):
    # S is a subset of Q
    return True
  # otherwise, S is not a subset of Q
  return False
```

```python
isSubset_bad({0, 7, 2}, {0, 2, 5})
```

```python
isSubset_bad(  {0, 2, 7} ,  {0, 2, 6} )
```

```python
isSubset_bad(  {0, 2, 7} ,  {0, 7, 2, 6} )
```

### Cartesian Product

```python
for a in {1, 2}:
  for b in {'a', 'b', 'c'}:
    print( (a, b) )
```

```python
{
  (a, b)
    for a in {1, 2}
    for b in {'a', 'b', 'c'}
}
```

```python
{
  (a, b)
    for a in range(7, 10) # a = 7, 8, 9
    for b in range(5, 10) # b = 5, 6, .., 9
    #if a > b
}
```

```python
def cartProd(A, B):
  return { (a, b)
              for a in A
              for b in B
        }
```

```python
cartProd( {1, 2}, {'a', 'b'} )
```

```python
cartProd( {1, 2}, {'a', 'b', 'd'} )
```
