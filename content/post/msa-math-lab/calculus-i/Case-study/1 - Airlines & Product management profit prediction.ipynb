## Case-Study: Predicting Which Air Line will be More Profitable

**Story.** You are given an offer to work in two international air lines companies: _A_ and _B_. You want to decide which of them is more profitable on both the short-term and long-term. Answer by the patterns of graphs you were studied.

```python
# Company A

days = np.arange(0, 5, 1)
profit = 1000 * days

plt.scatter( days, profit )
```

```python
# Company B

days = np.arange(0, 5, 1)
profit = 200 * (days**2)

plt.scatter( days, profit )
```

## Case-study: Which Product Manager is Profitable

**Story.** You are the CEO of an international air line company. You are given two plans by two product managers, Tameem and Antar, with plans increasing profits for the next year. Tameem will increase profits of each department by 1000. Antar will double each department.

For the following bar-charts, Which product manager will achieve more profit for the next year?

```python
department = ["A", "B", "C", "D", "E"]

# 10, 20, .., 50
profit = np.arange(10, 60, 10)

plt.title("January")
plt.bar(department, profit)
```

```python
department = ["A", "B", "C", "D", "E"]

# 10, 20, .., 50
profit = np.arange(1000, 6000, 1000)

plt.title("October")
plt.bar(department, profit)
```

#### Hints

- Try to sketch a function whose behavior is similar to the bar-chart.

#### Solution

Tameem for January and Antar for October.
