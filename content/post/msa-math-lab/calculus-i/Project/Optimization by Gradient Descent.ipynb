**Gradient Descent Algorithm**

````python
# Gradient descent for single variable
# Input: f'(x)
#        x_1
#        learning rate
#        threshold
#        maxIter
# Output: converged x-value, sequence of points (x_1,..,x_T), number of iterations
def gradDescSeq(df, x_prev, learn_rate, epsilon, maxIter):
  # initialize with first x
  sequence = np.array([x_prev])

  # Itterations number
  grad_iter = 1

  # First itteration of gradient descent
  x_next = x_prev - learn_rate*df(x_prev)

  # Save first itteration
  sequence = np.append(sequence, x_next)

  # while abs(x_{t+1} - x_{t}) > threshold, keep itterating gradient descent
  while ( abs(x_next - x_prev) > epsilon ) and ( grad_iter < maxIter ):

      # Update x_{t}
      x_prev = x_next

      # Gradient descent step
      # Set new x_{t+1} = x_{t} - lambda*f'(x_{t})
      x_next = x_prev - learn_rate*df(x_prev)

      # Save itteration
      sequence = np.append(sequence, x_next)

      # Increment itteration
      grad_iter += 1

  # Output converged x_{T}
  return x_next, sequence, grad_iter
````

### Error Function

Error at a point

````python
actual[0] - predictions[0]
````

````python
# absolute value
abs( predictions[0] - actual[0] )
````

There are plenty of error functions. A common standard one is $\sqrt{\sum (y_i - x_i)^2}$ but we do $\sum |y_i - x_i|$ as it's easier for students.

````python
def sumOfPointsError(actual, predictions):
  # a list for error at each point
  error_pnt = [ abs(actual[i] - predictions[i]) for i in range( len(actual) ) ]
  # summation of errors
  return np.sum(error_pnt)
````

````python
sumOfPointsError(actual, predictions)
````

````python
def linearPredictor(m, b, x):
  return (m*x) + b
````

Predictor of $m$ and $b$ yields an error

````python
# given m and b, construct the predictor, and compute predictions on x list
def computePredictions( m, b, x):
  return [ linearPredictor(m, b, i) for i in x ]
````

````python
# global variables
## x. list of points
## actual. actual customer budget
def errorOfPredictor( m, b ):
  # predictions of the linear model decided by m and b
  predictions = computePredictions( m, b, x)
  # error of the linear model decided by m and b
  return sumOfPointsError(actual, predictions)
````

````python
# input. symbolic expression
# output. pythonic function of it
def pyFunc(expr):
  return lambdify(x, expr)

# input. symbolic expression
# output. pythonic function of the derivative
def diffFunc(expr):
  # differentiate symbolically
  diff_expr = diff(expr, x)
  # pythonic function of the derivative
  return pyFunc(diff_expr)
````

Stepping was too high. We introduce _Learning rate_.

````python
learning_rate = 0.1
````
