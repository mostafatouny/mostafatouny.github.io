### Question

(ex. 32, p. 78, sec. 2.3, Briggs) Evaluate
\begin{align}
\lim_{h \rightarrow 0} \frac{3}{\sqrt{16 + 3h} + 4}
\end{align}

#### Solution

\begin{align}
&= \frac{\lim_{h \rightarrow 0} 3}{\sqrt{\lim_{h \rightarrow 0} 16 + \lim_{h \rightarrow 0} 3h} + \lim_{h \rightarrow 0} 4} \\
&= \frac{3}{\sqrt{16 + 3 \cdot 0} + 4} \\
&= \frac{3}{8}
\end{align}

### Question

(ex. 78, p.79, sec. 2.3, Briggs) Evaluate

\begin{align}
\lim_{x \rightarrow 1} \frac{x-1}{\sqrt{4x + 5} - 3}
\end{align}

#### Solution

\begin{align}
&\frac{x-1}{\sqrt{4x + 5} - 3} \\
&= \frac{x-1}{\sqrt{4x + 5} - 3} \cdot \frac{\sqrt{4x + 5} + 3}{\sqrt{4x + 5} + 3} \\
&= \frac{(x-1)(\sqrt{4x + 5} + 3)}{4(x-1)} \\
&= \frac{\sqrt{4x + 5} + 3}{4}
\end{align}
So
\begin{align}
\lim_{x \rightarrow 1} \frac{x-1}{\sqrt{4x + 5} - 3} &= \lim_{x \rightarrow 1} \frac{\sqrt{4x + 5} + 3}{4} \\
&= \frac{\sqrt{4(1) + 5} + 3}{4} \\
&= \frac{6}{4} = \frac{2}{3}
\end{align}

### Question

(p.74, Briggs). Compute
\begin{align}
\lim_{x \rightarrow 2} \frac{x^2 - 6x + 8}{x^2 - 4}
\end{align}

#### Solution

\begin{align}
\frac{x^2 - 6x + 8}{x^2 - 4} = \frac{(x-2)(x-4)}{(x-2)(x+2)} = \frac{x-4}{x+2}
\end{align}

It follows the intended limit is equal to
\begin{align}
\lim_{x \rightarrow 2} \frac{x-4}{x+2} = \frac{2-4}{2+2} = \frac{-2}{4} = -\frac{1}{2}
\end{align}



### Question

(ex. 37, p. 88, sec. 2.4, Briggs) Evaluate or show its divergence to infinity.

\begin{align}
\lim_{x \rightarrow 0^+} -10 \cot x
\end{align}

#### Solution

By definition
$$
\cot x = \frac{\cos x}{\sin x}
$$
But
\begin{align}
\lim_{x \rightarrow 0^+} \cos x &= 1 \\
\lim_{x \rightarrow 0^+} \sin x &= 0
\end{align}
Concluding
$$
\lim_{x \rightarrow 0^+} \frac{\cos x}{\sin x} = \infty
$$
Therefore
$$
\lim_{x \rightarrow 0^+} -10 \cdot \cot x = - \infty
$$

### Question

(ex. 34, p. 98, sec. 2.5, Briggs) Evaluate
$$
\lim_{x \rightarrow \infty} \frac{-x^3 + 1}{2x + 8}
$$

#### Solution

\begin{align}
&\frac{-x^3 + 1}{2x + 8} \\
&= \frac{-x^3 + 1}{2x + 8} \cdot \frac{1/x}{1/x} \\
&= \frac{-x^2 + 1/x}{2 + 8/x}
\end{align}
Therefore
\begin{align}
\lim_{x \rightarrow \infty} \frac{-x^3 + 1}{2x + 8} &= \lim_{x \rightarrow \infty} \frac{-x^2 + 1/x}{2 + 8/x} \\
&= -\infty
\end{align}
