### Question

(Thomas, Ex. 19, p. 39). Evaluate
\begin{align} \left ( \frac{2}{\sqrt{2}} \right )^4 \end{align}

#### Solution

\begin{align}
= \frac{2^4}{2^{\frac{1}{2} \cdot 4}} = \frac{2^4}{2^2} = 2^2 = 4
\end{align}

### Question

(Thomas, Ex. 3, p. 50). Decide whether the function $y = 2 |x|$ is one-to-one.

#### Hints

- (toolboxes) Recall the definition of injectivity from Rosen's discrete math.
- (decompose) Observe $|x| = |-x|$.
- (combine ideas) Formulate an algebraic equality that allows you to find a counter-example to injectivity.
- (toolbox) Proof by counter-example technique.

#### Solution

Counter-example. $f(1) = 2|1| = 2|-1| = f(-1)$. Hence, Not one-to-one.

### Question

(Thomas, Ex. 43, p. 51).
\begin{align}
a. \; e^{\ln 7.2} \quad b. \; e^{-\ln x^2} \quad c. \; e^{\ln x - \ln y}
\end{align}

#### Solution

**a.**
\begin{align}
= 7.2
\end{align}

**b.**
\begin{align}
&= e^{-2 \ln x} \\
&= \left ( e^{\ln x} \right )^{-2} \\
&= x^{-2} = \frac{1}{x^2}
\end{align}

**c.**
\begin{align}
&= e^{\ln x/y} = \frac{x}{y}
\end{align}

### Question

(Thomas, ex. 5, p. 19). Given $f(x) = x + 5$ and $g(x) = x^2 - 3$, find $f(g(x))$.

#### Solution

\begin{align}
= g(x) + 5 = x^2 - 3 + 5 = x^2 + 2
\end{align}

### Question

(advanced algebra, [6.5.3](https://math.libretexts.org/Bookshelves/Algebra/Advanced_Algebra/06%3A_Solving_Equations_and_Inequalities/605%3A_Solving_Quadratic_Inequalities)) Solve

\begin{align}
-x^2 + 6x + 7 \geq 0
\end{align}

#### Hints

- Set equality and solve for x.
- Graph the function, and observe how roots divide the function.
- Derive the general conclusion.

#### Solution

\begin{align}
-x^2 + 6x + 7 &= 0 \\
-(x^2 - 6x - 7) &= 0 \\
-(x+1)(x-7) &= 0 \\
\end{align}
Then $x = -1$ or $x = 7$. Moreover
\begin{align}
f(-3) &= -1(-2)(-10) \quad &\text{negative} \\
f(0) &= -1(1)(-7) \quad &\text{positive} \\
f(10) &= -1(11)(4) \quad &\text{negative}
\end{align}
So solution is $x \in [-1, 7]$
