**Shifting a sequence**

```python
# shifts the list's values in-place by c
def shiftListValues(x, c):
  for i in range(len(x)):
    x[i] = x[i] + c
```

```python
x = [1, 2, 3]
shiftListValues(x, 10)
x
```

**Testing symmetry on a sequence**

```python
# input: function f
#        list l of real numbers
# output: True if and only if f is even on all values of l
def testEvenFunc(f, l):
  for x in l:
    if not (f(x) == f(-x)):
      return True
  return False
```

```python
# f(x) = x^2
def f(x):
  return x^2

testEvenFunc(f, [1, 2, 3, 4, 5])
```
