## Helper Functions

```python
# @title
# Source: https://matplotlib.org/stable/gallery/showcase/integral.html

from matplotlib.patches import Polygon

def plotAreaUnderCurve(f, x, a, b):
  y = f(x)

  fig, ax = plt.subplots()
  ax.plot(x, y, 'b', linewidth=2)
  ax.set_ylim(bottom=0)

  # Make the shaded region
  ix = np.linspace(a, b)
  iy = f(ix)
  verts = [(a, 0), *zip(ix, iy), (b, 0)]
  poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
  ax.add_patch(poly)

  #ax.text(0.5 * (a + b), 30, r"$\int_a^b f(x)\mathrm{d}x$",
  #        horizontalalignment='center', fontsize=20)

  fig.text(0.9, 0.05, '$x$')
  fig.text(0.1, 0.9, '$y$')

  ax.spines[['top', 'right']].set_visible(False)
  ax.set_xticks([a, b], labels=[f'{a}', f'{b}'])
  #ax.set_yticks([])

  plt.show()
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

**Area Under Curve**

```python
# function
def f(x):
    return (x - 3) * (x - 5) * (x - 7) + 85

# integral limits
a, b = 2, 9

# sequence 0, 1, .., 10
x = np.linspace(0, 10)

plotAreaUnderCurve(f, x, a, b)
```

**Computing Area** ([source](https://docs.scipy.org/doc/scipy/tutorial/integrate.html))

```python
sp.integrate.quad(f, a, b)
```
