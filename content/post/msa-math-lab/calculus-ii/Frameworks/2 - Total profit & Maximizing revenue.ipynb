## Case-study: Total Profit for a Segment of Clients

**Story.** Your friend is a data analyst who works in an entertainment company. He shows you a graph of a distribution of clients. x-axis is the number of minutes they spend on an advertisement, and y-axis is the profit. He wants to know the total profit of clients spending 5 ~ 10 minutes.

You are only given a function `g(x)`:

```python
def g(x):
  return 1000*(1 - ( (0.2 * x) - 1 )**2)
```

### Solution

```python
sp.integrate.quad(g, 5, 10)
```

Total profit is 3333.

## Case-study: Maximizing Car Renting Revenue

(Ex. 4.35, [Sec. 4.7](https://openstax.org/books/calculus-volume-1/pages/4-7-applied-optimization-problems), Calculus I) Owners of a car rental company have determined that if they charge customers $p$ dollars per day to rent a car, where $50 \leq p \leq 200$, Then 1000 cars will be rented but each increment in price loses 5 cars. How much should they charge to maximize their revenue?


### Hint

- Formulation is a necessary step to use calculus toolboxes.
- Try graphing the function.
- Try a computational method.
- Spot a remarking property about the local maxima/minima.
- Remark a pattern correlating the sign of derivatives.

### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Formulation.** Number of cars to be rented is $c(p) = 1000 - 5 (p-50) $. Revenue is $r(p) = p \cdot c(p) = 1000p - 5p^2 + 250p = -5p^2 + 1250p$.

```python
def cars(p):
  return 1000 - (5*(p-50))

def revenue(p):
  return p * cars(p)
```

**Graphing.**

```python
# sequence 50, 51, .., 200
x = np.arange(50, 200)
# sequence of corresponding revenues
y = revenue(x)
```

```python
plt.plot( x, y)
```

**Analytic.** (Recall Calculus I)

\begin{align}
r'(p) &= 1250 - 10p \\
r'(p) &= 0 \\
p &= 125
\end{align}
