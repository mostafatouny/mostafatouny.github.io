Substituting `NULL`

```python
# every NULL is substituted by zero
df.fillna(0)
```

```python
# average order hour of day
avgHour = np.mean( df['order_hour_of_day'].to_numpy() )
# every NULL is substituted by the average
df = df.fillna(avgHour)
# error as the column `order_hour_of_day` is of Integer data type and `avgHour` is float
```

```python
# cast float as integer
df = df.fillna( int(avgHour) )
df
```
