**Story.** Your friend is a data analyst who works in an entertainment company. He shows you a graph of a distribution of clients. x-axis is the number of minutes they spend on an advertisement, and y-axis is the profit. He wants to know the total profit of clients spending 5 ~ 10 minutes.

You are only given a query `g(x)` for the profit at `minutes = x` which can be called as a black-box function, Due to a corruption in the company's database.

### Solution

```python
# @title
# Hidden function
# g(x) = 1 - ( (1/5)x - 1 )^2
def g(x):
  return 1000*(1 - ( (0.2 * x) - 1 )**2)
```

**Formulation.** The total profit is the area under an unknown function. It could be approximated by Riemann sum.

**Solution.** Using Riemann or Trapezoid methods for approximating the integral.

```python
x = np.linspace(5, 10, num=50)
y = g(x)
np.trapz(y, x)
```

So, Approximately the total profit is $3332$.
