**Single-variable function.** Table of values for the difference between two consequent points.

```python
import numpy as np
```

```python
x = np.arange(0.0, 5.0, 1.0)
y = 2 * x
```

```python
x
```

```python
y
```

```python
def rateOfChange( x1, y1, x2, y2 ):
  return (y2 - y1) / (x2 - x1)

def tableOfValues(x, y):
  n = len(x)
  for i in range(n-1):
    print( (y[i+1] - y[i]) / (x[i+1] - x[i]) )
```

```python
tableOfValues(x,y)
```

**Multi-variable function.** Table of values for the difference between two consequent points, Fixing the other axis.

```python
# f(a,b) = a^2 + b^2
def f(a,b):
  return a**2 + b**2

# fix f at a = 2
a_0 = 0

# construct sequence x = -5, -4, .., 5
x = np.arange(-5, 6)
# corresponding y, fixing a = 0
y = f(a_0, x)
#y = [ f(a_0, b) for b in x ]
```

```python
x
```

```python
y
```

```python
tableOfValues(x, y)
```
