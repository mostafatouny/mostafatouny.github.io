**Question.** (Thomas, sec. 5.2, ex. 39, p. 313).
- Find a formula for the Riemann sum obtained by dividing the interval into $n$ equal subintervals using the right-hand endpoint.
- Take the limit of these sums to calculate the area under the curve.

\begin{align}
f(x) = 1 - x^2 \quad x \in [0,1]
\end{align}

### Solution

\begin{align}
\Delta x &= \frac{1 - 0}{n} \\
c_i &= i \Delta x = \frac{i}{n} \\
g(n) &= \sum_{i=1}^n \Delta x \cdot f(c_i) \\
&= \sum_{i=1}^n \frac{1}{n} \cdot (1 - \left (\frac{i}{n} \right)^2) \\
&= \sum_{i=1}^n \frac{1}{n} \cdot \left (\frac{n^2}{n^2} - \frac{i^2}{n^2} \right ) \\
&= \frac{1}{n^3} \sum_{i=1}^{n} ( n^2 - i^2 ) \\
&= \frac{1}{n^3} \cdot \left [ n^3 - \sum_{i=1}^n i^2 \right ] \\
&= \frac{1}{n^3} \cdot \left [ n^3 - \frac{n(n+1)(2n+1)}{6} \right ] \\
&= 1 - \frac{n(n+1)(2n+1)}{6n^3} \\
&= 1 - \left [ \frac{1}{3} + \frac{1}{n} + \frac{1}{6n^2} \right ] \\
\lim_{n \rightarrow \infty} g(n) &= 1 - \left [ \frac{1}{3} + 0 + 0 \right ] = \frac{2}{3}
\end{align}
