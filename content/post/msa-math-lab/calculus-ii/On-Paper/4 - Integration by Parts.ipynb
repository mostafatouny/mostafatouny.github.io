### Question

(ex. 26, p. 513, sec. 7.2, Briggs) Evaluate

$$
A = \int x^2 \ln^2 x \, dx
$$

#### Hints

- You will use substitution by parts twice
- Set $u = ln^2 x$

#### Solution

\begin{align}
u &= \ln^2 x \quad du = 2 \frac{1}{x} \ln(x) \, dx \\
dv &= x^2 \, dx \quad v = \frac{1}{3} x^3
\end{align}

\begin{align}
A &= \frac{1}{3} x^3 \ln^2 x - \int \frac{1}{3} x^3 2 \frac{1}{x} \ln(x) \, dx \\
&= \frac{1}{3} x^3 \ln^2 x - \frac{2}{3} \int x^2 ln(x) \, dx
\end{align}

We now evaluate
$$
A' = \int x^2 ln(x) \, dx
$$

\begin{align}
u &= \ln x \quad du = \frac{1}{x} \, dx \\
dv &= x^2 \, dx \quad v = \frac{1}{3} x^3
\end{align}

\begin{align}
A' &= \frac{1}{3} x^3 \ln(x) - \int \frac{1}{3} x^3 \frac{1}{x} \, dx \\
&= \frac{1}{3} x^3 \ln(x) - \frac{1}{3} \int x^2 \frac{1}{x} \, dx \\
&= \frac{1}{3} x^3 \ln(x) - \frac{1}{3} \frac{1}{3} x^3 \\
&= \frac{1}{9} x^3 (3 \ln(x) - 1) + C
\end{align}

It follows
\begin{align}
A &= \frac{1}{3} x^3 \ln^2(x) - \frac{2}{3} \cdot \frac{1}{9} x^3 (3\ln(x) - 1) \\
&= \frac{x^3}{27} (9 \ln^2 x - 6 \ln x + 2) + C
\end{align}

### Question

(ex. 36, p. 513, sec. 7.2, Briggs) Evaluate

$$
A = \int_0^{\frac{1}{\sqrt{2}}} x \tan^{-1}(x^2) \, dx
$$


#### Hints

- You will integrate twice, the second one by substitution.
- Recall $[ tan^{-1}(x) ]' = \frac{1}{1+x^2}$, p. 213 in Briggs.
- Set $u(x) = \tan^{-1}(x^2)$.

#### Solution

\begin{align}
u(x) &= \tan^{-1}(x^2) \quad u'(x) = \frac{2x}{1 + x^4} \, dx \\
v'(x) &= x \, dx \quad v(x) = \frac{1}{2} x^2
\end{align}

\begin{align}
A = tan^{-1}(x^2) \cdot \frac{1}{2} x^2 \bigg |_0^{1/\sqrt{2}} - \int_0^{1/\sqrt{2}} \frac{x^3}{1 + x^4} \, dx
\end{align}

We now evaluate by substitution
$$
A' = \int_0^{1/\sqrt{2}} \frac{x^3}{1 + x^4} \, dx
$$

$$
u = 1 + x^4 \quad du = 4x^3 \, dx
$$

\begin{align}
A' &= \frac{1}{4} \int_1^{5/4} \frac{1}{u} \, du \\
&= \frac{1}{4} \ln(u) \bigg |_1^{5/4} \\
&= \frac{1}{4} (\ln(5/4) - ln(1)) \\
&= \frac{1}{4} \ln(5/4)
\end{align}

It follows
\begin{align}
A &= \left [ tan^{-1}(1/2) \frac{1}{2} \frac{1}{2} - 0 \right ] - \frac{1}{4} \ln(5/4) \\
&= \frac{1}{4} \left [ \tan^{-1}(1/2) - \ln(5/4) \right ]
\end{align}
