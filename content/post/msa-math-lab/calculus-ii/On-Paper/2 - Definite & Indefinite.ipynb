### Question

(Thomas, Section 5.3, Ex. 13, p. 322). Assume $\int_0^3 f(z) \, dz = 3$ and $\int_0^4 f(z) \, dz = 7$. Find $\int_3^4 f(z) \, dz$.

#### Hints

- (Decompose components; Think of the problem differently) Consider the numbers $0$, $3$, and $4$. Form an algebraic equality.
- (Combine ideas) Use the additivity rule in _table 5.4_, p. 317.

#### Solution

\begin{align}
\int_0^3 f(z) \, dz + \int_3^4 f(z) \, dz &= \int_0^4 f(z) \, dz \\
3 + \int_3^4 f(z) \, dz &= 7 \\
\int_3^4 f(z) \, dz &= 7 - 3 = 4
\end{align}

### Question

(Thomas, Section 4.8, Ex. 41, p. 285). Find
$$ \int \frac{t \sqrt{t} + \sqrt{t} }{t^2}  \, dt $$

#### Hints

- (Trick) Simplify the algebraic form.
- See _Rules for Exponents_ in _Thomas_, p. 37.

#### Solution

\begin{align}
f(t) &= \frac{t \cdot t^{1/2} + t^{1/2}}{t^2} \\
&= \frac{t \cdot t^{1/2}}{t^2} + \frac{t^{1/2}}{t^2} \\
&= t^{-1/2} + t^{-3/2}
\end{align}
By linearity (Table 4.3, p. 281),
\begin{align}
\int f(t) \, dt &= 2t^{1/2} + -2 t^{-1/2} + c \\
&= 2 \sqrt{t} - 2 \frac{1}{\sqrt{t}} + c
\end{align}

### Question

(Briggs, Section 5.3, Ex. 42, p. 367). Compute

\begin{align}
\int_0^\pi (1 - \sin x) \, dx
\end{align}

#### Hints

- Recall the integral rules of trigonometric functions. See _Theorem 3.12_, p. 163.
- Recall properties of definite integrals. See _Table 5.4_, p. 349.
- Recall the definition of trigonometric functions. See _Figure 1.62_, p. 39.

#### Solution

\begin{align}
\int_0^\pi (1 - \sin x) \, dx = x + \cos x \bigg |_0^\pi = (\pi + \cos \pi) - (0 + \cos 0) = (\pi - 1) - (1) = \pi - 2
\end{align}
