## Question

(ex. 11, p. 659, sec. 9.4, Briggs) Decide whether the series converges.

$$
\sum_{k=2}^\infty  \frac{k}{\ln k}
$$

### Hints

- Use divergence test in page 648.
- You will use L'Hopital.

### Solution

\begin{align}
\lim_{n \rightarrow \infty} \frac{n}{\ln n} &= \lim_{n \rightarrow \infty} \frac{[n]'}{[\ln n]'} \\
&= \lim_{n \rightarrow \infty} \frac{1}{1/n} \\
&=\lim_{n \rightarrow \infty} n \\
&= \infty
\end{align}

Hence by divergence test, page 648, It diverges.

## Question

(ex. 21, p. 659, sec. 9.4, Briggs) Decide whether the series converges.

$$
\sum_{k=1}^\infty k e^{-2k^2}
$$

### Hints

- Use Integral test, p. 651.
- Compute the indefinite integral, then evaluate the limit of the integral.

### Solution

Indefinite integral

$$
A = \int ke^{-2k^2}
$$

By substitution,
\begin{align}
u &= -2k^2 \\
du &= -2 \cdot 2 \cdot k \, dk \\
&= -4 k \, dk
\end{align}

\begin{align}
A &= \frac{-1}{4} \int -4 k e^{-2k^2} \, dk \\
&= \frac{-1}{4} \int e^u \, du \\
&= \frac{-1}{4} e^u + C \\
&= \frac{-1}{4} e^{-2k^2} + C
\end{align}

It follows
\begin{align}
&\int_1^\infty ke^{-2k^2} \\
&= \lim_{R \rightarrow \infty} \left ( \frac{-1}{4} e^{-2R^2} - \frac{-1}{4} e^{-2(1)^2} \right ) \\
&= 0 - \frac{-1}{4} e^{-2(1)^2} \\
&= \frac{1}{4e^2}
\end{align}

Converges by the integral test, p. 651.

## Question

(ex. 31, 658, sec. 9.4, Briggs) Decide whether the series converges.

$$
\sum_{k=3}^\infty \frac{1}{(k-2)^4}
$$

### Hints

- Shift the series so that $k$ starts from $1$.

### Solution

$$
\sum_{k=3}^\infty \frac{1}{(k-2)^4} = \sum_{k=1}^\infty \frac{1}{(k)^4}
$$

Since $k > 1$, By the _p-series test_, It converges.
