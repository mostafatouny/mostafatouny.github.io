### Question

(ex. 7, p. 529, sec. 7.4, Briggs) Evaluate
$$
A = \int_0^{5/2} \frac{dx}{\sqrt{25 - x^2}}
$$

#### Solution

\begin{align}
x &= 5 \sin \theta \\
dx &= 5 \cos \theta \, d\theta \\
\sqrt{25 - x^2} &= \sqrt{5^2 - x^2} = \sqrt{5^2 - (5 \sin \theta)^2} = \sqrt{5^2 (1 - \sin^2 \theta)} = \sqrt{5^2 \cos^2 \theta} = 5 \cos \theta \\
\frac{5}{2} = 5 \sin \theta &\rightarrow \theta = \frac{\pi}{6} \\
0 = 5 \sin \theta &\rightarrow \theta = 0 \\
A &= \int_0^{\pi/6} \frac{5 \cos \theta}{5 \cos \theta} \, d\theta \\
&= \int_0^{\pi/6} d\theta \\
&= \theta \bigg |_0^{\pi/6} = \frac{\pi}{6} - 0 = \frac{\pi}{6}
\end{align}

### Question

(ex. 13, p. 529, sec. 7.4, Briggs) Evaluate
$$
A = \int \frac{1}{(16-x^2)^{1/2}} \, dx
$$

#### Solution

\begin{align}
\sqrt{16-x^2} &= \sqrt{4^2-x^2} = \sqrt{a^2-x^2} \\
a &= 4 \\
x &= a \sin \theta = 4 \sin \theta \\
dx &= 4 \cos \theta \, d \theta \\
\sqrt{4^2 - x^2} &= \sqrt{4^2 - (4 \sin \theta)^2} = \sqrt{4^2 (1-sin^2 \theta)} = \sqrt{4^2 \cos^2 \theta} = 4 \cos \theta \\
A &= \int \frac{4 \cos \theta}{4 \cos \theta} \, d \theta \\
&= \int 1 \, d \theta \\
&= \theta + C \\
&= \sin^{-1} \left (\frac{x}{4} \right ) + C
\end{align}

### Question

(ex. 21, p. 529, sec. 7.4, Briggs) Evaluate
$$
\int \frac{1}{x^2 \sqrt{x^2 + 9}} \, dx
$$

#### Solution

\begin{align}
\sqrt{9+x^2} &= \sqrt{3^2+x^2} = \sqrt{a^2 +x^2} \\
a &= 3 \\
x &= a \tan \theta = 3 \tan \theta \\
dx &= 3 \sec^2 \theta \, d \theta \\
\sqrt{x^2 + 9} &= \sqrt{(3 \tan \theta)^2 + 9} = \sqrt{3^2 \sec^2 \theta} = |3 \sec \theta | = 3 \sec \theta \\
A &= \int 3 \sec^2 \theta \cdot \frac{1}{3^2 \tan^2 \theta} \cdot \frac{1}{3 \ sec \theta} \, d \theta \\
&= \int \sec \theta \cdot \frac{1}{3^2 \tan^2 \theta} \, d \theta \\
&= \frac{1}{9} \cdot \int \frac{1}{\cos \theta} \cdot \frac{\cos^2 \theta}{\sin^2 \theta} \\
&= \frac{1}{9} \cdot \int \cos \theta \cdot \frac{1}{\sin^2 \theta} \, d \theta \\
u &= \sin \theta \\
du &= \cos \theta \, d \theta \\
A &= \frac{1}{9} \int \frac{1}{u^2} \, du \\
&= \frac{1}{9} \cdot -1 \cdot u^{-1} \\
&= \frac{-1}{9} \cdot \frac{1}{\sin \theta} \\
&= \frac{-\sqrt{x^2 + 9}}{9x}
\end{align}

### Question

(ex. 31, p. 529, sec. 7.4, Briggs) Evaluate
$$
\int \frac{\sqrt{x^2 - 9}}{x} \, dx \quad x > 3
$$

#### Solution

\begin{align}
\sqrt{x^2 - 9} &= \sqrt{x^2 - 3^2} = \sqrt{x^2 - a^2} \\
a &= 3 \\
x &= 3 \sec \theta \\
dx &= 3 \sec \theta \tan \theta \\
A &= \int 3 \tan \theta \cdot \frac{1}{3 \sec \theta} \cdot 3 \sec \theta \, d \theta \\
&= 3 \int \tan^2 \theta \, d \theta \\
&= 3 \cdot  \left (\frac{\tan \theta}{1} - \theta \right ) + C \\
&= 3 \cdot \left [ \frac{\sqrt{x^2 - 3^2}}{3} - sec^{-1}(x/3) \right ] + C \\
&= \sqrt{x^2 - 9} - 3 \sec^{-1}(x/3) + C
\end{align}
