**Certifying logical equivalence by trying all assignments**

```python
def osama(p, q):
  #if ( not (p and q) ) == ( (not p) or (not q) ):
  #  return True
  #return False
  return ( not (p and q) ) == ( (not p) or (not q) )
```

```python
osama(True, True)
```

```python
osama(False, True)
```

```python
for p in [True, False]:
  for q in [True, False]:
    print( osama(p, q) )
```

**Satisfiability**

Under all possible assignments, If the statement in one of them is True, It is satisfiable. If under all possible assignments, the statement is False, Then it is not satisfiable.

```python
def st(p, q):
  return p and q
```

```python
# initially the flag is flase
flag = False
# try all possible assignments
for p in [True, False]:
  for q in [True, False]:
    # if the statement is true in one of them
    if st(p,q) == True:
      # statement is satisfiable
      flag = True
print(flag)
```

```python
def st_1(p, q):
  return (p and not p) and q
```

```python
# initially the flag is flase
flag = False
# try all possible assignments
for p in [True, False]:
  for q in [True, False]:
    # if the statement is true in one of them
    if st_1(p,q) == True:
      # statement is satisfiable
      # however, that line is never called as the statement is unsatisfiable
      flag = True
# flag value remains the same, False
print(flag)
```
