```python
# Input. Matrix of connections. 2d list
#        User index. integer
# Output. True iff User is connected to somebody on Facebook
def isConnected(M, k):
  # loop on kth row
  for i in M[k]:
    # if connected to somebody
    if i == 1:
      return True
  # if loop is finished, and we are here
  # then no one is found connected
  return False
```

```python
isConnected(M, 0)
```

## Recursive Function - Rosen 5.3

Defining a Function as an assignment

Domain: Students, $\{ Arwa, Mohannad \}$

Co-domain: Integers

$f = \{  (Arwa, 2), (Mohannad, 4) \}$

$f(Arwa) = 2$

$f(Mohannad) = 4$

Defining a function by algebra

Domain: Real numbers $R$

Range: Real numbers $R$

$f: R \rightarrow R$

$f(x) = x^2$

$f(0) = 0$

$f(3) = 9$

$f(-2) = 4$

$(0,0) \in f$

$(3, 9) \in f$

$(-2, 4) \in f$

Defining a function recursively

Example 1 (page 367)

$f(0) = 3$

$f(1) = 2f(0) + 3 = 9$

$f(2) = 2 f(1) + 3 = 2(9) + 3 = 21$

..etc

for any $f(n)$ where $n = 0, 1, .. $

```python
# as code
def f(n):
  if n == 0:
    return 3
  return (2 * f(n-1)) + 3
```

```python
# compute f(2)
# n = 2
# 2 * f(1) + 3
# compute f(1)
# n = 1
# 2 * f(0) + 3
# compute f(0)
# return 3
# return 2*3 + 3 = 9
# return 2*9 + 3 = 21
```
