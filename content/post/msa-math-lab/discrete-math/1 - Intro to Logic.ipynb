**AND**

```python
def my_AND_0(p, q):
  if p == True and q == True:
    return True
  if p == True and q == False:
    return False
  if p == False and q == True:
    return False
  if p == False and q == False:
    return False
```

```python
my_AND_0(True, False)
```

```python
my_AND_0(True, True)
```

**XOR**

```python
def my_XOR_0(p, q):
  if p and q:
    return False
  if p and not q:
    return True
  if not p and q:
    return True
  if not p and not q:
    return False
```

```python
def my_XOR_1(p, q):
  if p == q:
    return False
  return True
```

```python
my_XOR_1(False, False)
```

```python
def my_XOR_2(p, q):
  return not (p == q)
```

```python
my_XOR_2(False, True)
```

**Conditional Statement**

$p$: It is raining <br>
$q$: I will go shopping <br>
Is $p \rightarrow q$ True, Given $p = False$, In other words,

$q$: I will buy you a car <br>
Is $p \rightarrow q$ True, Given $p = False$, In other words,

Actual: Arwa got less than 3.5 GPA, and Her dad DID NOT buy her a car, Is her Dad truthful? YES

Actual: Arwa got less than 3.5 GPA, and Her dad bought her a car, Is her Dad truthful? YES, i.e $p \rightarrow q = True$

Actual: Arwa got at least 3.5 GPA, and Her dad bought her a car, Is her Dad truthful? YES, i.e $p \rightarrow q = True$

Actual: Arwa got at least 3.5 GPA, and Her dad did NOT buy her a car, Is her Dad truthful? NO, i.e $\neg (p \rightarrow q)$

```python
def my_imp_0(p, q):
  if p and not q:
    return False
  return True
```

**Contrapositive**

When is $p \rightarrow q$ False? only if $p = True$ and $q = False$.

When is $\neg q \rightarrow \neg p$ False? Only if $\neg q = True$ and $\neg p = False$. Equivalently, $q = False$ and $p = True$.

Hence both are logically equivalent.

**Biconditional Statement**

$p \leftrightarrow q$ is a shorthand of (logically equivalent to) $p \rightarrow q \wedge q \rightarrow p$.

```python
def my_biCond_0(p, q):
  return my_imp_0(p, q) and my_imp_0(q, p)
```
