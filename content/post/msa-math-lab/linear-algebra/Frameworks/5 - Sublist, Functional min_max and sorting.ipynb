Finding the minimum number in a list

```python
minimum = l[0]
for i in l:
  if i < minimum:
    minimum = i
minimum
```

```python
min(l)
```

```python
# minimum as a function of foo
def foo(x):
  return abs(x - 8)
```

```python
minimum = foo(l[0])
for i in l:
  if foo(i) < minimum:
    minimum = foo(i)
minimum
```

```python
# minimum element, by foo function
minimum = foo(l[0])
el = l[0]
for i in l:
  if foo(i) < minimum:
    minimum = foo(i)
    el = i
el
```

```python
min(l, key = foo)
```

```python
min(l, key = lambda x: abs(x-8) )
```
