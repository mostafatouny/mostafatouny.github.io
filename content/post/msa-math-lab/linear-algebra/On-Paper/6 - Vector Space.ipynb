### Question

(ex. 12 (a), p. 187, sec. 4.1, Kolman) Given

$$
u = \begin{bmatrix} 2 \\ 3 \end{bmatrix}
\quad
v = \begin{bmatrix} -2 \\ 5 \end{bmatrix}
$$

Compute $u + v$, $u - v$, $2u$, and $3u-2v$.

#### Solution

\begin{align}
u + v &= \begin{bmatrix} 2 + (-2) \\ 3 + 5 \end{bmatrix} = \begin{bmatrix} 0 \\ 8 \end{bmatrix} \\
u - v &= \begin{bmatrix} 2 - (-2) \\ 3 - 5 \end{bmatrix} = \begin{bmatrix} 4 \\ -2 \end{bmatrix} \\
2u &= \begin{bmatrix} 2 \cdot 2 \\ 2 \cdot 3 \end{bmatrix} = \begin{bmatrix} 4 \\ 6 \end{bmatrix} \\
3u - 2v &= \begin{bmatrix} 3 \cdot 2 \\ 3 \cdot 3 \end{bmatrix} - \begin{bmatrix} 2 \cdot (-2) \\ 2 \cdot 5 \end{bmatrix} \\
&= \begin{bmatrix} 6 \\ 9 \end{bmatrix} - \begin{bmatrix} -4 \\ 10 \end{bmatrix} \\
&= \begin{bmatrix} 6 - (-4) \\ 9 - 10 \end{bmatrix} = \begin{bmatrix} 10 \\ -1 \end{bmatrix}
\end{align}
