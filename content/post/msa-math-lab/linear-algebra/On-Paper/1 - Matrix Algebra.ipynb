### Question

(Kolman, Ex. 14(f), p. 31). Given
\begin{align}
A = \begin{bmatrix}
1 & 2 & 3 \\
2 & 1 & 4
\end{bmatrix}
\quad
C = \begin{bmatrix}
3 & -1 & 3 \\
4 & 1 & 5 \\
2 & 1 & 3
\end{bmatrix}
\quad
E = \begin{bmatrix}
2 & -4 & 5 \\
0 & 1 & 4 \\
3 & 2 & 1
\end{bmatrix}
\end{align}
Compute $A(C - 3E)$

#### Solution

\begin{align}
3E &= \begin{bmatrix}
6 & -12 & 15 \\
0 & 3 & 12 \\
9 & 6 & 3
\end{bmatrix} \\
C - 3E &= \begin{bmatrix}
-3 & 11 & -12 \\
4 & -2 & -7 \\
-7 & -5 & 0
\end{bmatrix} \\
A(C - 3E) &= \begin{bmatrix}
1 \cdot -3 + 2 \cdot 4 + 3 \cdot -7 & 1 \cdot 11 + 2 \cdot -2 + 3 \cdot -5 & 1 \cdot -12 + 2 \cdot -7 + 3 \cdot 0 \\
2 \cdot -3 + 1 \cdot 4 + 4 \cdot -7 & 2 \cdot 11 + 1 \cdot -2 + 4 \cdot -5 & 2 \cdot -12 + 1 \cdot -7 + 4 \cdot 0
\end{bmatrix} \\
&= \begin{bmatrix}
-16 & -8 & -26 \\
-30 & 0 & -31
\end{bmatrix}
\end{align}
