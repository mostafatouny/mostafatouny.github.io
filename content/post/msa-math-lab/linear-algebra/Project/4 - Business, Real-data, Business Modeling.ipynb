## Linear-algebraic Modeling

**Movies' Features as a Matrix**

Columns: `filmtv_id`, `avg_vote`,	`critics_vote`,	`public_vote`,	`total_votes`

\begin{bmatrix}
2 & 7.7 &	8.00 & 7.0 & 22 \\
3 & 6.5 &	6.00 & 7.0 & 4 \\
17 & 5.7 & 6.00	& 5.0	& 10
\end{bmatrix}

**User's Feature as a Matrix**

Columns
- `user_id`, `Gender`,	`spotify_usage_period`,`Spotify_listening_device`, `spotify_subscription_plan`,	`premium_sub_willingness`

Data
- User 0: Female,	More than 2 years, Smart speakers or voice assistants, Free (ad-supported), Yes
- User 1: Male, More than 2 years, Computer or laptop, Free (ad-supported), Yes

Modeling
- Categorical (Nominal) data, Labeled as 0, 1, 2, ..
  - Male: 1, Female: 0
  - Smartphone: 0, Computer or laptop: 1, Smart speakers or voice assistants: 2

Matrix
\begin{bmatrix}
0 & 2 & 2 & 0 & 1 \\
1 & 2 & 1 & 0 & 1
\end{bmatrix}





**User-Movie Matrix**

Movies
- 0: Oppenheimer
- 1: Barbie
- 2: The origin of species

Users
- 0: Antar
- 1: Tameem
- 2: Logain

Watching Record
- Antar saw only Oppenheimer
- Tameem saw only Oppenheimer and The origin of species
- Logain saw Barbie and The origin of species

Matrix (Rows: users, Columns: Movies)
\begin{bmatrix}
1 & 0 & 0 \\
1 & 0 & 1 \\
0 & 1 & 1
\end{bmatrix}

A matrix cell represents a user interaction with the movie.
